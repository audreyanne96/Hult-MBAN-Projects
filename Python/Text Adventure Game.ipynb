```python
"""
    DocString:
    
    A) INTRODUCTION:
    This game is based on the TV sitcom How I Met Your Mother (HIMYM) which ran from 2005-2014. 
    Set in NYC, the show is about Ted Mosby telling his kids how he met their mother.
    This is also the main premise of this game -- i.e., finding the mother.
    
    B) GAMEPLAY:
    There are 5 stages in total. Each stage has 2-3 trivia questions about the show. Answering the
    questions correctly will increase the chances of the Ted meeting The Mother. There are also
    questions in most stages intended for Ted to choose a path to move forward. Answering those
    questions incorrectly will trigger a fail() wherein the game will end.
    
    C) SUMMARY OF STAGES:
    Stage1: The Apartment
    - 2 questions for points
    - 1 question with fail()
    Stage2: MacLaren's
    - 2 questions for points
    - 1 question with fail()
    Stage3: St. Patrick's Day Bar
    - 3 questions for points
    Stage4: Farhampton
    - 3 questions for points
    - 4 questions with fail()
    Stage5: Farhampton Train Station
    - Final stage for results
"""
# SETTING UP VARIABLES
# Importing Packages
import random
import time

# Setting-up variables for scoring
points_scored = 0 # Temporary variable to store points scored for a question
prob_scored = 0 # Temporary variable to store percentage increase in chance of meeting the mother
points_total = 0 # Temporary variable to store cumulative points scored
prob_total = 0 # Temporary variable to store cumulative percentage chance of meeting the mother
result = 0 # Temporary variable to store final result: True = Meets The Mother, False = Does not meet The Mother
game = True # Game status: True = continue, False = End
umbrella = "" # Temporary variable to store status of getting the umbrella. Important in Stage 3
bar_rob = "" # Temporary variable to store choice between going to Barney or Robin.

# Setting up dictionary with points for trivia questions
# Code is <stage> questions for points <question number>
# e.g., s1qp1 = stage 1, question for points 1
points = dict(s1qp1=dict(a=5,b=0,c=5),
              s1qp2=dict(a=5,b=5,c=0),
              s2qp1=dict(a=0,b=10,c=0),
              s2qp2=dict(a=10,b=0,c=0),
              s3qp1=dict(a=0,b=0,c=10),
              s3qp2=dict(a=10,b=0,c=0),
              s3umb=dict(a=110,b=0), # special question that will grant 55% increase in chance
              s4qp1=dict(a=10,b=0,c=0),
              s4qp2=dict(a=0,b=10,c=0),
              s4qp3=dict(a=10,b=0,c=0))

# Scoring Function
# Will increase points_total depending on the question and answer the player chooses
def scoring(question,choice):
    
    global points_scored
    global prob_scored
    global points_total
    global prob_total
    global points

    # Will store the points and percentage scored for one question
    points_scored = points[question][choice]
    prob_scored = round((points_scored/200)*100,0)
    
    # Will store the cumulative points and percentage scored for one question
    points_total += points[question][choice]
    prob_total = round((points_total/200)*100,0)
    
    return points_scored
    return prob_scored
    return points_total
    return prob_total
    
# Answer Response Function (MAIN)
# Will let the player know how many points were scored and running total.
def answer_response(choice):
    
    global points_scored
    global prob_scored
    global points_total
    global prob_total

    # Will inform the player if the answer was correct or incorrect,
    # How many points and % chance was earned and the cumulative % the player has scored.
    choice = choice.upper()
    # If the answer is correct (i.e., the player scored points in that question)
    if points_scored > 0:
        print(f"""
        {"*" * 50}
        You answered {choice} which is CORRECT.
        You have increased your chances of meeting The Mother by {prob_scored}%.
        You now have a {prob_total}% chance of meeting The Mother.
        {"*" * 50}
        """)
    # If the answer is incorrect (i.e., the player did not score points in that question)
    elif points_scored == 0:
        print(f"""
        {"*" * 50}
        You answered {choice} which was INCORRECT.
        You are still at {prob_total}% chance of meeting The Mother.
        {"*" * 50}
        """)
    else:
        print(f"""
        Error in answer_response().
        """)
# Answer Response Function (SPECIAL)
# Answer response specifically for the Stage 3 during the umbrella question
# Will let the player know how many points were scored and running total.
def answer_response_umb(choice):
    
    global points_scored
    global prob_scored
    global points_total
    global prob_total

    # Will inform the player if the answer was earned them points and % chance or not
    choice = choice.upper()
    # If the answer is correct (i.e., the player scored points in that question)
    if points_scored > 0:
        print(f"""
        {"*" * 50}
        You answered {choice} increasing your chances of meeting The Mother by {prob_scored}%.
        You now have a {prob_total}% chance of meeting The Mother.
        {"*" * 50}
        """)
    # If the answer is incorrect (i.e., the player did not score points in that question)
    elif points_scored == 0:
        print(f"""
        {"*" * 50}
        You answered {choice} keeping you at {prob_total}% chance of meeting The Mother.
        {"*" * 50}
        """)
    else:
        print(f"""
        Error in answer_response_umb.
        """)   

# Determining if Ted meets the mother
# Using a code similar to a biased coinflip to determine if ted meets the mother
# If True = Meets Mother
# If False = Does not meet mother
def final():
    
    global result
    global prob_total
    
    # Random.random() will generate a float number between 0 and 1.
    # If this number is less than prob_total, it will be True
    # Otherwise it will be False
    # As such, the higher prob_total is, the higher chance to be True
    if random.random() < prob_total: 
        result = True
        return result
    else:
        result = False
        return result        

# Fail Function
# Will trigger the end of the game if the wrong path is chosen
# The response will depend on which stage or question the fail was triggered
def fail(ft):
    
    global game
    global prob_total
    
    fail_type = ft.casefold()
    # Code corresponds to the function where the fail was triggered
    # Code: stage <stage number> question with fail <question number, if applicable>
    # Stage 1
    if fail_type == "s1qf":
        print(f"""
        {"*"*50}
        FINAL
        {"-"*20}
        You have chosen to stay in and stay in you will.
        Although you earned {prob_total}% chance to meet The Mother, you chose the wrong path.
        You have missed the chance to meet your destiny, The Mother of your children.
        You will be single... forever...
        {"*"*50}
        """)
    # Stage 2
    elif fail_type == "s2qf":
        print(f"""
        {"*"*50}
        FINAL
        {"-"*20}
        You have chosen to wait for Marshall and Lily.
        They don't end up coming down. You go back to your apartment.
        Although you earned {prob_total}% chance to meet The Mother, you chose the wrong path.
        You have missed the chance to meet your destiny, The Mother of your children.
        You will be single... forever...
        {"*"*50}
        """)
    # Stage 4, question with fail 1
    elif fail_type == "s4qf1":
        print(f"""
        {"*"*50}
        FINAL
        {"-"*20}
        You say your goodbyes to Marshall and Lily. And you leave a note for Barney and Robin.
        You go to the train station to catch a train back to NY, and move to Chicago.
        Although you earned {prob_total}% chance to meet The Mother, you chose the wrong path.
        You have missed the chance to meet your destiny, The Mother of your children.
        You will be single... forever...
        {"*"*50}
        """)
    # Stage 4, question with fail 2/3 - barney
    elif fail_type == "s4qf23b":
        print(f"""
        {"*"*50}
        FINAL
        {"-"*20}
        Barney talks to Robin, they cancel the wedding. 
        You tell them the truth about moving to Chicago. 
        You still chose to go despite the wedding getting cancelled. 
        You go to the train station to catch a train back to NY.
        Although you earned {prob_total}% chance to meet The Mother, you chose the wrong path.
        You have missed the chance to meet your destiny, The Mother of your children.
        You will be single... forever...
        {"*"*50}
        """)
    # Stage 4, question with fail 2/3 - robin
    elif fail_type == "s4qf23r":
        print(f"""
        {"*"*50}
        FINAL
        {"-"*20}
        Robin cries but you says your goodbyes. The wedding happens but you don't attend. 
        You go to the train station to catch a train back to NY.
        Although you earned {prob_total}% chance to meet The Mother, you chose the wrong path.
        You have missed the chance to meet your destiny, The Mother of your children.
        You will be single... forever...
        {"*"*50}
        """)
    # Stage 4, question with fail 4
    elif fail_type == "s4qf4":
        print(f"""
        {"*"*50}
        FINAL
        {"-"*20}
        Marshall and Lily cry. They say their goodbyes. The wedding happens without you. 
        You go to the train station to catch a train back to NY.
        Although you earned {prob_total}% chance to meet The Mother, you chose the wrong path.
        You have missed the chance to meet your destiny, The Mother of your children.
        You will be single... forever...
        {"*"*50}
        """)
    else:
        print(f"""
        Error in fail().
        """)
    
    # Whenever a fail is triggered, it will automatically stop the game
    # game = False = game will not proceed to next stage
    game = False
    return game

# INTRO FUNCTIONS
# Will welcome the player into the game and set the scene and storyline
def intro():
    
    print(f"""
    {"-"*90}
    Welcome to How I Met Your Mother: Find The Mother Edition.
    
    You will be playing as Ted Mosby, a hopeless romantic looking for "The One."
    You live in the most amazing city in the world - New York City, in an apartment the
    viewers of this show are surprised you can afford. You live with your best friend from college
    and his wife - Marshall and Lily. Again, viewers are surprised that this is not awkward at all.
    
    Dating has not been very kind to you. Even with your wingman... and best friend (cause
    he'll get mad otherwise)... the legen- wait for it -dary Barney, it has been largely
    unsuccessful. 
    
    Until you met Robin. From casually dating to being friends to being best friends to finally 
    seriously dating, the past couple of years were great. But then, reality caught up, and it was
    clear you wanted different things.
    
    You are now back in the dating scene. You are now again trying to find "The One".
    {"-"*90}
    """)

# STAGE 1 FUNCTIONS
# Setting: The Apartment
# Win: Choose to go to MacLaren’s
# Fail: Stay at home

# Stage 1 question for points 1
def s1qp1():
    
    try:
        s1qp1_choice = input(f"""
        What is Lily painting?
        A. A painting of an elephant
        B. The gang hanging out at a coffee shop
        C. A colorful interpretation of the NYC skyline, freaking out a bird passing by
        Answer: """)
        s1qp1_choice = s1qp1_choice.casefold()
        while s1qp1_choice != "a" and s1qp1_choice != "b" and s1qp1_choice != "c":
            s1qp1_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s1qp1_choice == "a" or s1qp1_choice == "b" or s1qp1_choice == "c":
                break
        scoring("s1qp1",s1qp1_choice)
        answer_response(s1qp1_choice)
    except:
        print(f"""
        Error in s1qp1.
        """)
    
# Stage 1 question for points 2
def s1qp2():
    
    try:
        s1qp2_choice = input(f"""
        What is Marshall doing?
        A. Studying for the NY state bar exam
        B. Inventing Marshgammon
        C. Making a trifle
        Answer: """)
        s1qp2_choice = s1qp2_choice.casefold()
        while s1qp2_choice != "a" and s1qp2_choice != "b" and s1qp2_choice != "c":
            s1qp2_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s1qp2_choice == "a" or s1qp2_choice == "b" or s1qp2_choice == "c":
                break
        scoring("s1qp2",s1qp2_choice)
        answer_response(s1qp2_choice)
    except:
        print(f"""
        Error in s1qp2.
        """)

# Stage 1 question with fail
def s1qf():
    
    global game
    try:
        s1qf_choice = input("""
        What do you do?
        A. Go to MacLaren’s
        B. Stay in for the night
        Answer: """)
        s1qf_choice = s1qf_choice.casefold()
        while s1qf_choice != "a" and s1qf_choice != "b":
            s1qf_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s1qf_choice == "a" or s1qf_choice == "b":
                break
        if s1qf_choice == "a":
            print("""
    Ted \t: Fine. But you better come down to MacLaren’s. 
            \t  I don’t wanna get stuck with wingman-ing Barney for the whole night.
            """)
            game = True
            return game
        elif s1qf_choice == "b":
            print("""
    Ted \t: Nah. I’m just gonna stay in. 
            \t  It’s just gonna be another night of Barney flirting with every female at the bar.
            """)
            fail("s1qf")
    except:
        print(f"""
        Error in s1qf.
        """)
        
# Bringing together Stage 1
def stage1():
    
    print(f"""
    THE APARTMENT
    {"-"*20}
    It's just another Friday night. You just got back from work and Lily and Marshall are
    in the living room. You look at Lily and find her in the middle of painting, while 
    Marshall is doing what he is best at in the couch.
    """)
    
    time.sleep(3)
    s1qp1()
    time.sleep(3)
    s1qp2()
    time.sleep(3)

    print("""
    Lily \t: Oh hey, Ted! How was work?
    Ted \t: Well...
    Marshall \t: That bad, huh? Wanna grab some drinks?
    Ted \t: I don’t know, man.
    Lily \t: Barney’s been calling us to get down to MacLaren’s. You should go ahead.
    Marshall \t: Yeah, man. We’ll be going down soon too when Lily is done.
    Ted \t: Will Robin be there?
    Lily \t: Ted, it’s still awkward for her too. You aren’t the only one having a hard time.
    Ted \t: *sighs*
    """)
    
    time.sleep(5)
    s1qf()

# STAGE 2 FUNCTIONS
# Setting: MacLaren's
# Win: Choose to celebrate St. Patrick's Day with Barney at a Club
# Fail: Go home

# Stage 2 question for points 1
def s2qp1():
    
    try:
        s2qp1_choice = input(f"""
        But he notices you and then, before you can even get a drink, you hear him say to the woman:
        A. How you doin'?
        B. Have you met, Ted?
        C. That's my spot.
        Answer: """)
        s2qp1_choice = s2qp1_choice.casefold()
        while s2qp1_choice != "a" and s2qp1_choice != "b" and s2qp1_choice != "c":
            s2qp1_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s2qp1_choice == "a" or s2qp1_choice == "b" or s2qp1_choice == "c":
                break
        scoring("s2qp1",s2qp1_choice)
        answer_response(s2qp1_choice)
    except:
        print("""
        Error in s2qp1.
        """)

# Stage 2 question for points 2
def s2qp2():
    
    try:
        s2qp2_choice = input(f"""
        Left in an awkward situation, you start talking to the woman. You find out her name is:
        A. Trudy
        B. Janice
        C. Sherry
        Answer: """)
        s2qp2_choice = s2qp2_choice.casefold()
        while s2qp2_choice != "a" and s2qp2_choice != "b" and s2qp2_choice != "c":
            s2qp2_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s2qp2_choice == "a" or s2qp2_choice == "b" or s2qp2_choice == "c":
                break
        scoring("s2qp2",s2qp2_choice)
        answer_response(s2qp2_choice)
    except:
        print("""
        Error in s2qp2.
        """)

# Stage 2 question with fail
def s2qf():
    
    global game
    try:
        s2qf_choice = input("""
        What do you do?
        A. Stay at MacLaren’s and wait for Lily and Marshall
        B. Go with Barney to another bar to celebrate Saint Patrick’s Day
        Answer: """)
        s2qf_choice = s2qf_choice.casefold()
        while s2qf_choice != "a" and s2qf_choice != "b":
            s2qf_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s2qf_choice == "a" or s2qf_choice == "b":
                break
        if s2qf_choice == "b":
            print("""
    Ted \t: You know what, Barney? This might be exactly what I need.
    Barney \t: YESSSSS. LET'S GO.
            """)
            game = True
            return game
        elif s2qf_choice == "a":
            print("""
    Ted \t: You know what, Barney? This really isn’t what I need right now.
    Barney \t: FIIIIINE. Whatever.
            """)
            fail("s2qf")
            game = False
            return game
    except:
        print("""
        Error in s2qf.
        """)

# Bringing together Stage 2
def stage2():
    
    print(f"""
    MACLAREN'S
    {"-"*20}
    You get to the bar and see Barney. Of course, he is at the bar flirting with a beautiful woman. 
    You approach Carl to order a drink, trying to stay out of Barney’s radar.
    """)
    
    time.sleep(3)
    s2qp1()
    time.sleep(3)
    s2qp2()
    time.sleep(3)
        
    print("""
    You finish speaking to the woman and approach Barney at your regular booth.
    
    Ted \t: Thank you for the warning there hm?
    Barney \t: As your best friend…
    Ted \t: You aren’t…
    Barney \t: AS YOUR BEST FRIEND, I know what you need. And, Ted, what you need tonight is...
    Ted \t: A drink and a good night’s sle——
    Barney \t: A DRINK AND A GOOD NIGHT OUT FOR SAINT PATRICK’S DAY!
    Ted \t: Barney, no. Marshall and Lily are coming down here and that’s all I need tonight.
    Barney \t: *shock* 
    Barney \t: Are you... Am I hearing this right? Is my best friend...
    Ted \t: I'm not your best friend.
    Barney \t: IS MY BEST FRIEND turning down a night out in the best city in the world?
    Ted \t: I’ll be fine.
    Barney \t: Come on, Ted. Tonight will be legen- wait for it - dary!
    """)
    
    time.sleep(5)
    s2qf()

# STAGE 3 FUNCTIONS
# Setting: St. Patrick's Day Club
# Win: Get additional points for getting the umbrella
# Fail: No fail

# Stage 3 question for points 1
def s3qp1():
    
    try:
        s3qp1_choice = input(f"""
        Because you are a good friend, you remember:
        A. A bro must always buy two pizzas.
        B. A bro must always paint a mural for their bro’s kids’ nursery.
        C. A bro will, in a timely manner, alert his bro to the existence of a girl fight.
        Answer: """)
        s3qp1_choice = s3qp1_choice.casefold()
        while s3qp1_choice != "a" and s3qp1_choice != "b" and s3qp1_choice != "c":
            s3qp1_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s3qp1_choice == "a" or s3qp1_choice == "b" or s3qp1_choice == "c":
                break
        scoring("s3qp1",s3qp1_choice)
        answer_response(s3qp1_choice)
    except:
        print(f"""
        Error in s3qp1.
        """)

# Stage 3 question for points 2
def s3qp2():
    
    try:
        s3qp2_choice = input(f"""
        You try to run away, but then Barney says something and you now have to fight because:
        A. A bro must always oblige to fulfill the actions of a double-dog dare.
        B. A bro must never run away.
        C. A bro must never wear a raw turkey on their head to scare their bro.
        Answer: """)
        s3qp2_choice = s3qp2_choice.casefold()
        while s3qp2_choice != "a" and s3qp2_choice != "b" and s3qp2_choice != "c":
            s3qp2_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s3qp2_choice == "a" or s3qp2_choice == "b" or s3qp2_choice == "c":
                break
        scoring("s3qp2",s3qp2_choice)
        answer_response(s3qp2_choice)
    except:
        print("""
        Error in s3qp2.
        """)

# Stage 3 question for points - SPECIAL
def s3umb():
    
    global umbrella
    try:
        s3umb_choice = input(f"""
        What do you do?
        A. Grab an umbrella and go home
        B. Wait out the rain and go home
        Answer: """)
        s3umb_choice = s3umb_choice.casefold()
        while s3umb_choice != "a" and s3umb_choice != "b":
            s3umb_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s3umb_choice == "a" or s3umb_choice == "b":
                break
        scoring("s3umb",s3umb_choice)
        answer_response_umb(s3umb_choice)
        if s3umb_choice == "a":
            umbrella = True
            return umbrella
        else:
            umbrella = False
            return umbrella
    except:
        print("""
        Error in s3umb.
        """)

# Bringing together Stage 3
def stage3():
    
    print(f"""
    ST. PATRICK'S DAY CLUB
    {"-"*20}
    Ted and Barney go to a club celebrating St. Patrick’s Day. They immediately realize that 
    the universe is on their side. They ditch their dates and were rewarded by being able to
    cut the line to get in the club. They order 2 bottles of champagne, on someone else’s 
    tab, and get rewarded with free caviar. Tonight is their night.
    
    Suddenly, there was a commotion going on in the dance floor. 
    """)
    
    time.sleep(3)
    s3qp1()
    time.sleep(3)
    
    print(f"""
    Things are starting to get crazy as the night goes on. All of a sudden, your luck
    started to turn around. The owner of the tab you ordered 2 expensive champagnes on 
    found out and is looking to fight you.""")
    
    time.sleep(3)
    s3qp2()
    time.sleep(3)
    
    print(f"""
    Well, you got knocked out. Obviously. You wake up in the emergency room, with the nurse 
    judging your choices heavily. You ask for your phone and find out you did not have it 
    with you when you got sent to the ER. It is turning into a very bad day, the complete 
    opposite of the previous night.
    
    When you got out of the hospital, you go back to the bar to look for your phone. You even 
    feel the weather getting worse. Rain is coming.
    
    The bar is a mess and unfortunately you don’t find your phone. Even worse, the rain started 
    and is becoming more terrible by the second. Another bad day.
    """)
    
    time.sleep(5)
    s3umb()

# STAGE 4 FUNCTIONS
# Setting: Farhamption
# Win: Give Robin and Barney your blessing and attend the wedding ceremony
# Fail 1: Don’t talk to robin or barney.
# Fail 2/3: Don’t give robin or barney your blessing, don’t attend the ceremony
# Fail 3: Give robin and barney your blessing but don’t attend the ceremony

# Stage 4 question for points 1
def s4qp1():
    
    try:
        s4qp1_choice = input(f"""
        What bank was this ATM for?
        A. Goliath National Bank
        B. Bank of America
        C. Harrington Private Bank
        Answer: """)
        s4qp1_choice = s4qp1_choice.casefold()
        while s4qp1_choice != "a" and s4qp1_choice != "b" and s4qp1_choice != "c":
            s4qp1_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s4qp1_choice == "a" or s4qp1_choice == "b" or s4qp1_choice == "c":
                break
        scoring("s4qp1",s4qp1_choice)
        answer_response(s4qp1_choice)
    except:
        print(f"""
        Error in s4qp1.
        """)

# Stage 4 question for points 2
def s4qp2():
    
    try:
        s4qp2_choice = input(f"""
        Where is Ted moving to?
        A. Pasadena, CA
        B. Chicago, IL
        C. Boston, MA
        Answer: """)
        s4qp2_choice = s4qp2_choice.casefold()
        while s4qp2_choice != "a" and s4qp2_choice != "b" and s4qp2_choice != "c":
            s4qp2_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s4qp2_choice == "a" or s4qp2_choice == "b" or s4qp2_choice == "c":
                break
        scoring("s4qp2",s4qp2_choice)
        answer_response(s4qp2_choice)
    except:
        print(f"""
        Error in s4qp2.
        """)

# Stage 4 question for points 3
def s4qp3():
    
    try:
        s4qp3_choice = input(f"""
        What building was demolished to give way for Ted's building?
        A. The Arcadian
        B. The Ashburn
        C. The Ascalla
        Answer: """)
        s4qp3_choice = s4qp3_choice.casefold()
        while s4qp3_choice != "a" and s4qp3_choice != "b" and s4qp3_choice != "c":
            s4qp3_choice = input(f"""
            Kindly only input A, B or C.
            Answer: """)
            if s4qp3_choice == "a" or s4qp3_choice == "b" or s4qp3_choice == "c":
                break
        scoring("s4qp3",s4qp3_choice)
        answer_response(s4qp3_choice)
    except:
        print("""
        Error in s4qp3.
        """)

# Stage 4 question with fail 1
def s4qf1():

    global game
    global bar_rob
    
    try:
        s4qf1_choice = input("""
        What do you do?
        A. Go to Barney
        B. Go to Robin
        C. Say your goodbyes to Marshall and Lily. Leave a note for Barney and Robin then leave.
        Answer: """)
        s4qf1_choice = s4qf1_choice.casefold()
        while s4qf1_choice != "a" and s4qf1_choice != "b" and s4qf1_choice != "c":
            s4qf1_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s4qf1_choice == "a" or s4qf1_choice == "b":
                break
        if s4qf1_choice == "a":
            print("""
    Ted \t: I’ll go and speak to Barney.
            """)
            bar_rob = "barney"
            game = True
            return game
            return bar_rob
        elif s4qf1_choice == "b":
            print("""
    Ted \t: I’ll head to Robin's room.
            """)
            bar_rob = "robin"
            game = True
            return game
            return bar_rob
        elif s4qf1_choice == "c":
            print(f"""
    Ted \t: I... Could you give this note to both of them please? I love them both. 
        \t  I love you guys. But, I can't be here.
            """)
            fail("s4qf1")
            game = False
            return game
            return bar_rob
    except:
        print(f"""
        Error in s4qf1.
        """)                

# Stage 4 question with fail 2/3 - Barney's Room
def s4qf23_barney():
    
    global game
    
    print(f"""
    Ted goes to Barney’s room.
    
    Barney \t: Ted! Ted! You’re here. Thank god, you’re here. I have a problem.
    Ted \t: Woah. Calm down first. What is it?
    Barney \t: SUITS! I CAN’T CHOOSE A SUIT! NOTHING FITS.
    Ted \t: But you have a million suits. You wear one everyday!
    Barney \t: I don’t know what’s going on, Ted. Everything is wrong. Nothing fits, nothing 
        \t  looks good. Maybe Robin and I shouldn’t get married. Is this a sign, Ted?
    """)
    
    time.sleep(3)
    
    try:
        s4qf23b_choice = input(f"""
        What do you do?
        A. Tell Barney that it is a sign and he should talk to Robin about it.
        B. Tell Barney that he’s just nervous, and he just needs to think of his love for Robin.
        """)
        s4qf23b_choice = s4qf23b_choice.casefold()
        while s4qf23b_choice != "a" and s4qf23b_choice != "b":
            s4qf23b_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s4qf23b_choice == "a" or s4qf23b_choice == "b":
                break
        if s4qf23b_choice == "a":
            print(f"""
    Ted \t: Barney, if you are thinking that this small problem is a sign, maybe you 
        \t  should talk to Robin about getting married.
    Barney \t: You’re right, Ted. Maybe I was never meant for marriage.
            """)
            fail("s4qf23b")
            game = False
            return game
        elif s4qf23b_choice == "b":
            print(f"""
    Ted \t: Barney, think about Robin. Just think about her. You love her, don’t you?
    Barney \t: Yes, very much.
    Ted \t: Then it doesn’t matter if the suit doesn’t fit. What matters is you
        \t  are marrying the love of your life.
    Barney \t: *Barney’s suit suddenly fits* 
    Barney \t: You are right, Ted. This is right. This feels right.
            """)
            game = True
            return game
    except:
        print("""
        Error in s4qf23_barney.
        """)        

# Stage 4 question with fail 2/3 - Robin's Room
def s4qf23_robin():
    
    global game
    
    print(f"""
    Ted goes to Robin’s room.
    
    Robin \t: Ted? Is that you?
    Ted \t: Hey. You look beautiful.
    Robin\t: Thank you. I know it is difficult for you, but thank you for being here today.
    """)
    
    time.sleep(3)
    
    try:
        s4qf23r_choice = input(f"""
        What do you do?
        A. Tell Robin it's too difficult. You still love her and you just came to say goodbye.
        B. Tell Robin it's difficult but you're happy for them. You tell her about moving.
        """)
        s4qf23r_choice = s4qf23r_choice.casefold()
        while s4qf23r_choice != "a" and s4qf23r_choice != "b":
            s4qf23r_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s4qf23r_choice == "a" or s4qf23r_choice == "b":
                break
        if s4qf23r_choice == "a":
            print(f"""
    Ted \t: It is. And, I’m sorry Robin, but I can’t stand by and watch you get married.
    Robin \t: What? Why?
    Ted \t: It’s always been you. I still love you, Robin. And because of that, I can’t stay. 
        \t  I just came to see you cause I think you deserve to get it from me. 
        \t  I’m moving to Chicago.
    Robin \t: What?! You’re moving away? I know this is difficult for our friendship, 
        \t  but do you have to move away?
    Ted \t: I think we both know we’ve always needed some distance. New York is just 
        \t  filled with memories of us. I need to do this for me.
            """)
            fail("s4qf23r")
            game = False
            return game
        elif s4qf23r_choice == "b":
            print(f"""
    Ted \t: It is difficult but I’m happy for both of you.
    Robin \t: Thank you, Ted. You don’t know how much your blessing means to me.
    Ted \t: That being said though, I can’t stay in New York. I’m moving to Chicago.
    Robin \t: You’re moving away? I know this is difficult for our friendship, but
        \t  do you have to move away?
    Ted \t: I think we both know we’ve always needed some distance. New York is just 
        \t  filled with memories of us. I need to do this for me.
            """)
            game = True
            return game
    except:
        print(f"""
        Error in s4qf23_robin.
        """)
        
# Stage 4 question with fail 4
def s4qf4():

    global game
    
    try:
        s4qf4_choice = input("""
        What do you do?
        A. Attend the ceremony and leave after the reception.
        B. Don’t attend the ceremony and leave.
        Answer: """)
        s4qf4_choice = s4qf4_choice.casefold()
        while s4qf4_choice != "a" and s4qf4_choice != "b":
            s4qf4_choice = input(f"""
            Kindly only input A or B.
            Answer: """)
            if s4qf4_choice == "a" or s4qf4_choice == "b":
                break
        if s4qf4_choice == "a":
            print("""
    Ted \t: Yes. More than anything, they are my best friends. I want to be there for them. 
        \t  This is important for them.
            """)
            game = True
            return game
        elif s4qf4_choice == "b":
            print("""
    Ted \t: Yes, I don’t think I can do this. Will you please tell them congratulations for me?
    Marshall \t: Okay man.
    Marshall \t: *hugs Ted*
    Ted \t: I’ll miss you guys. When the time is right, I’ll come visit.
    Lily \t: Bye, Ted.
            """)
            fail("s4qf4")
            game = False
            return game
    except:
        print("""
        Error in s4qf4.
        """)        

# Bringing together Stage 4 part 1
def stage4_1():
    
    print(f"""
    Years has passed. A lot has happened. Both good and bad.
    
    You got engaged but got left at the altar. You had your building design chosen, but the 
    project got scrapped. You became a professor, but taught the wrong class on your
    first day - Econ 305.
    
    Now though, you face the most difficult choice.
    
    After getting together, breaking up and then getting together again, Robin is marrying Barney. 
    This is painful for you to see as you’ve always thought Robin was “The One” despite everything.
    You still have feelings for her, and you always will. It feels like there’s unfinished business
    and so seeing her marry Barney hurts.
    
    Oh... also, Marshall and Lily have a baby now. They're love life is fine. As usual.
    *cue the eyeroll*""")
    
    time.sleep(10)
    
    print(f"""
    FARHAMPTON
    {"-"*20}
    It's the day of the wedding. You just arrived in Farhampton and you see Lily over at an ATM.
    """)
    
    time.sleep(1)
    s4qp1()
    time.sleep(3)
    
    print(f"""
    Lily sees you and rushes towards you very worried.
    
    Lily \t: Are you sure you are really doing this Ted? I know this isn’t a happy moment for you, 
         \t  but that doesn’t mean you need to move thousands of miles away. You’re a New Yorker 
         \t  through and through.
    Ted \t: Yeah, Lily. But what am I supposed to do? I think it’s time for me to get 
        \t  the distance that I’ve always needed to move on.
    Lily \t: Just… at least say goodbye before you go.
    """)
    
    time.sleep(3)
    s4qp2()
    time.sleep(3)
    
    print(f"""
    Marshall arrives with their son, Marvin.
    
    Marshall \t: Tell me Lily lied to me.
    Ted \t: I… I can’t stay, Marshall. After everything, it’s just too difficult now.
    Marshall \t: You were the lead architect of the latest addition to the New York City skyline. 
        \t  There’s no way you can’t get a new projects here instead of moving to Chicago. 
        \t  You fought so hard for this, from the demolition to the construction. 
        \t  Why leave now?
    """)
    
    time.sleep(3)
    s4qp3()
    time.sleep(3)
    
    print(f"""
    Ted \t: I just need a fresh start. And the offer the Chicago firm gave me was great.
    Marshall \t: Could you at least talk to Barney and Robin first? Beyond anything 
        \t  that happened, we’re all friends. They deserve to know about this.
    """)
    
    time.sleep(3)
    s4qf1()

# Bringing together Stage 4 parts 2 and 3
def stage4_23():

    global bar_rob
    global game
    
    if bar_rob == "barney":
        s4qf23_barney()
        time.sleep(3)
        if game == True:
            robin_choice = input("""
            What do you want to do now?
            A. Go to Robin's room.
            B. Wait for the ceremony.
            Answer: """)
            robin_choice = robin_choice.casefold()
            while robin_choice != "a" and robin_choice != "b":
                robin_choice = input(f"""
                Kindly only input A or B.
                Answer: """)
                if robin_choice == "a" or robin_choice == "b":
                    break
            if robin_choice == "a":
                s4qf23_robin()
                time.sleep(3)
    elif bar_rob == "robin":
        s4qf23_robin()
        time.sleep(3)
        if game == True:
            barney_choice = input("""
            What do you want to do now?
            A. Go to Barney's room.
            B. Wait for the ceremony.
            Answer: """)
            barney_choice = barney_choice.casefold()
            while barney_choice != "a" and barney_choice != "b":
                barney_choice = input(f"""
                Kindly only input A or B.
                Answer: """)
                if barney_choice == "a" or barney_choice == "b":
                    break
            if barney_choice == "a":
                s4qf23_barney()
                time.sleep(3)
    else:
        print(f"""
        Error in stage4_23.
        """)

# Bringing together Stage 4 part 4
def stage4_4():
    
    print(f"""
    The ceremony is about to start. You are outside the ballroom with Marshall and Lily.
    
    Lily \t: This is finally happening.
    Marshall \t: You are amazing for being here, Ted. But you know you don’t have too right? 
        \t  We all know this is difficult for you.
    """)
    
    time.sleep(3)
    s4qf4()

# STAGE 5 FUNCTIONS
# No questions, just presentation of final results

# When result = True, umbrella = True
def rTuT():
    
    global points_total
    global prob_total
    global result
    
    print(f"""
    It's raining... You arrive at the station, and see a beautiful girl at the 
    train platform (with a familiar umbrella). And then you had the guts to stand up, 
    walk over, tap her on the shoulder, open your mouth and speak. You recognize her
    as part of the wedding band.
    
    Ted \t: Excuse me.
    Woman \t: Hi!
    Ted \t: Hi. I'm...
    Woman \t: The best man. Oh... Get in here.
    Ted \t: Great show tonight.
    Woman \t: Thank you.
    Ted \t: You're Cindy's ex-roommate right?
    Woman \t: And you're the professor. I took one of your classes.
    Ted \t: Really? Which one?
    Woman \t: Econ 305.
    Ted \t: Econ 305? I don't teach Econ 305... Oh no...
    Woman \t: Oh yeah.
    Ted \t: Excuse me, I'm going to jump into the tracks now.
    Woman \t: No, no, no. Don't. *smiles* You were great.
    Ted \t: Wait a second. This is my umbrella. I left this umbrella at Cindy's. 
        \t  You totally stole my umbrella.
    Woman \t: What? No, I didn't. This is my umbrella. I bought this.
    Ted \t: Excuse me, it even has my initials on here. "T.M.". Ted Mosby.
    Woman \t: Uh huh. Look again Ted Mosby. Those are my initials. "T.M.". Tracy McConnell.
    Ted \t: Uhm no, Tracy McConnell. It's "T.M.", Totally My umbrella.
    Tracy \t: You're "T.M.", Terribly Mistaken because this umbrella has always 
        \t  belonged "T.M.", To Me.
    Tracy \t: *laughs* 
    Tracy \t: Although, I did lose it for a few years there. So I went to this dance club...
    Ted \t: On St. Patrick's Day.
    Tracy \t: On St. Patrick's Day.
    Ted \t: And you left it there.
    Tracy \t: And I left it there.
    Ted \t: And you never thought you'd see it again.
    Tracy \t: And I never thought I'd see it again.
    *long pause*
    Tracy \t: Funny how sometimes, you just... find things.
    """)
    
    time.sleep(10)
    
    print(f"""
    {"*" * 50}
    FINAL
    {"-"*20}
    You earned {points_total} out of 200 points.
    This gave you a {prob_total}% chance of going to the train station at the right time.
    And you did! Just the right time...
    By also choosing the correct path, you met Tracy McConnell, The (future) Mother of your children.
    Congratulations! You are no longer single!
    {"*" * 50}
    """)

# When result = True, umbrella = False
def rTuF():
    
    global points_total
    global prob_total
    global result
    
    print(f"""
    It's raining... You arrive at the station, and see a beautiful girl at the 
    train platform. And then you had the guts to stand up, walk over, tap her on the shoulder, 
    open your mouth and speak. You recognize her as part of the wedding band.
    
    Ted \t: Excuse me.
    Woman \t: Hi!
    Ted \t: Hi. I'm...
    Woman \t: The best man. Oh... Get in here.
    Ted \t: Great show tonight.
    Woman \t: Thank you.
    Ted \t: You're Cindy's ex-roommate right?
    Woman \t: And you are the professor. I took one of your classes.
    Ted \t: Really? Which one?
    Woman \t: Econ 305.
    Ted \t: Econ 305? I don't teach Econ 305... Oh no...
    Woman \t: Oh yeah.
    Ted \t: Excuse me, I'm going to jump into the tracks now.
    Woman \t: No, no, no. Don't. *smiles* You were great.
    """)

    time.sleep(5)
    
    print(f"""
    {"*" * 50}
    FINAL
    {"-"*20}
    You earned {points_total} out of 200 points.
    This gave you a {prob_total}% chance of going to the train station at the right time.
    And you did! Just the right time... (Phew! Talk about being lucky!)
    By also choosing the correct path, you met Tracy McConnell, The (future) Mother of your children.
    Congratulations! You are no longer single!
    {"*" * 50}
    """)

# When result = False, umbrella = False
def rFuF():
    
    global points_total
    global prob_total
    global result
    
    print(f"""
    It's raining... You arrive at the station, and see that you are alone.
    """)
    
    time.sleep(1)
    
    print(f"""
    {"*" * 50}
    FINAL
    {"-"*20}
    You earned {points_total} out of 200 points.
    This gave you a {prob_total}% chance of going to the train station at the right time.
    Unfortunately, you chose to go to the station too early. The party was still going and the music rumbling.
    Despite choosing the right path, you did not meet Tracy McConnell, The (future) Mother of your children.
    You will be single... forever...
    {"*" * 50}
    """)

# When result = False, umbrella = True
def rFuT():
    
    global points_total
    global prob_total
    global result
    
    print(f"""
    It's raining... You arrive at the station, and see that you are alone.
    """)
    
    time.sleep(1)
    
    print(f"""
    {"*" * 50}
    FINAL
    {"-"*20}
    You earned {points_total} out of 200 points.
    This gave you a {prob_total}% chance of going to the train station at the right time.
    Unfortunately, you chose to go to the station too early. The party was still going and the music rumbling.
    Despite choosing the right path, you did not meet Tracy McConnell, The (future) Mother of your children.
    You will be single... forever...
    {"*" * 50}
    """)

# Bringing together Stage 5
def stage5():
    
    global points_total
    global umbrella
    global result
    
    print(f"""
    FARHAMPTON TRAIN STATION
    {"-"*20}
    Despite the difficult situation, the ceremony and the reception was beautiful. The music, 
    especially, was wonderful. Ted spoke to Marshall, Lily, Robin and Barney for a final time. 
    They all congratulated him on his new job, but were sad it meant he had to move away. 
    Ted goes to the train station.
    """)
    
    time.sleep(3)
    final()
    time.sleep(3)

    if result == True and umbrella == True:
        rTuT()
    elif result == True and umbrella == False:
        rTuF()
    elif result == False and umbrella == True:
        rFuT()
    elif result == False and umbrella == False:
        rFuF()

# GAME PROPER

# Bringing together the full game
# The game will continue to the next stages/parts so long as the game variable remains True
def game():
    
    intro()
    time.sleep(10)
    stage1()
    if game == True:
        time.sleep(3)
        stage2()
        if game == True:
            time.sleep(3)
            stage3()
            time.sleep(3)
            stage4_1()
            if game == True:
                time.sleep(3)
                stage4_23()
                if game == True:
                    time.sleep(3)
                    stage4_4()
                    if game == True:
                        time.sleep(3)
                        stage5()

# Will run when the fail() is triggered or when the player finishes the game
    print(f"""
    {"-"*40} THE END {"-"*40}
    """)

# RUNNING THE GAME
game()
```
