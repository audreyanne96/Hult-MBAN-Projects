# Problem Statement

### Problem 1 - Regional Market Research

Upon exploring all of the data provided, the set of data from the World Bank showed the most information for all countries and regions around the world. The World Bank data was not complete. However, for the analysis, only indices and metrics relevant to determining *Leisure Time*, *Market Size*, and *Market Stability* were necessary. - Specific indices and metrics taken from these databases were also chosen in consideration of the completeness of the information for all countries.

>The above files were then cleaned and merged. Various dataframes were created in preparation for later analyses.

```python
!pip install chart_studio
!pip install -U kaleido
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
import seaborn as sns
import zipfile as zf
from chart_studio import plotly as py
from scipy import stats
import kaleido
import warnings
warnings.filterwarnings('ignore')

sns.set_palette("Set3")

# ACCESSING WORLD BANK INFORMATION
# r = raw df for the file, will remain untouched

# Accessing: zip file for World Bank data
zf_wb = zf.ZipFile("data/World Bank.zip")
# Accessing: Countries and Regions
r_wb_country = pd.read_csv(zf_wb.open("world-development-indicators/Country.csv"),low_memory=False)
# Accessing: File for Healthcare Indices
r_wb_healthcare = pd.read_csv(zf_wb.open("countries-dataset-2020/Health care index by countries 2020.csv"),low_memory=False)
# Accessing: File for Quality of Life Index
r_wb_qol = pd.read_csv(zf_wb.open("countries-dataset-2020/Quality of life index by countries 2020.csv"),low_memory=False)
# Accessing: Files for Human Development Index to be used in determining Leisure Time
r_wb_deved = pd.read_csv(zf_wb.open("list_of_developed_countries.csv"),low_memory=False)
r_wb_deving = pd.read_csv(zf_wb.open("list_of_developed_or_developing_countries.csv"),low_memory=False)
r_wb_udev = pd.read_csv(zf_wb.open("list_of_less_or_under_developed_countries.csv"),low_memory=False)
# Accessing: Files for 2020 population and year-on-year change
r_wb_pop2020 = pd.read_csv(zf_wb.open("population-by-country-2020/population_by_country_2020.csv"),low_memory=False)
# Accessing: File for % of middle-aged (15-64) population
r_wb_agedist = pd.read_csv(zf_wb.open("countries-dataset-2020/Coutries age structure.csv"),low_memory=False)
# Accessing: File for Local Purchasing Power Index and Cost of Living Index
r_wb_col = pd.read_csv(zf_wb.open("2020-cost-of-living/cost of living 2020.csv"),low_memory=False)

# CLEANING: Main file for countries and regions
wb_country = r_wb_country[["Alpha2Code",
                           "ShortName",
                           "LongName",
                           "Region"]]
wb_country = wb_country.rename(columns={"Alpha2Code":"country_code",
                                        "ShortName":"country",
                                        "LongName":"country_full",
                                        "Region":"country_region"})
wb_country = wb_country.dropna(subset=["country_region"]) # Dropping rows related to aggregations with no associated regions
wb_country["country_code"] = wb_country["country_code"].str.strip()
wb_country["country"] = wb_country["country"].str.strip()
wb_country["country_full"] = wb_country["country_full"].str.strip()
wb_country["country_region"] = wb_country["country_region"].str.strip().astype("category")
wb_country["country_lowercase"] = wb_country["country"].str.lower()

# CLEANING: Main DF for Leisure Time (LT) Indices
# Cleaning for Health Care Index and Healthcare Index (Exponential)
lt_hci = r_wb_healthcare[["Country",
                          "Health Care Index",
                          "Health Care Exp. Index"]]
lt_hci = lt_hci.rename(columns={"Country":"country",
                                "Health Care Index":"lt_hci",
                                "Health Care Exp. Index":"lt_hciexp"})
lt_hci["country_lowercase"] = lt_hci["country"].str.strip().str.lower()
lt_hci = lt_hci.drop(columns=["country"])
lt_hci = lt_hci.groupby("country_lowercase",as_index=False).mean()
lt_hci = lt_hci.drop_duplicates(subset=["country_lowercase"])
# Cleaning for Quality of Life Index
lt_qol = r_wb_qol[["Country",
                   "Quality of Life Index"]]
lt_qol = lt_qol.rename(columns={"Country":"country",
                                "Quality of Life Index":"lt_qol"})
lt_qol["country_lowercase"] = lt_qol["country"].str.strip().str.lower()
lt_qol = lt_qol.drop(columns=["country"])
lt_qol = lt_qol.groupby("country_lowercase",as_index=False).mean()
lt_qol = lt_qol.drop_duplicates(subset=["country_lowercase"])
# Cleaning for Human Development Index
lt_hdi = pd.concat([r_wb_deved,r_wb_deving,r_wb_udev])
lt_hdi = lt_hdi.rename(columns={"Country Name":"country",
                                "Human Development Index":"lt_hdi"})
lt_hdi["country_lowercase"] = lt_hdi["country"].str.strip().str.lower()
lt_hdi = lt_hdi.drop(columns=["country",
                              "GNI per capita",
                              "Human Asset Index",
                              "Population 2020",
                              "Development Status"])
lt_hdi = lt_hdi.groupby("country_lowercase",as_index=False).mean()
lt_hdi = lt_hdi.drop_duplicates(subset=["country_lowercase"])
# Merging all indicators for Leisure Time
lt_indicators_df = [lt_hci,lt_qol,lt_hdi]
leisuretime_main = wb_country
for df in lt_indicators_df:
    leisuretime_main = pd.merge(leisuretime_main,
                           df,
                           how="left",
                           on="country_lowercase")
# Filling null values using median number per region
lt_indicators_col = ["lt_hci","lt_hciexp","lt_qol","lt_hdi"]
for column in lt_indicators_col:
    leisuretime_main[column] = leisuretime_main.groupby("country_region")[column].transform(lambda x:x.fillna(x.median()))
    leisuretime_main[column] = leisuretime_main[column].round(4)
# Preparing a trimmed DF for merging later in the analysis
lt_trim = leisuretime_main[["country_lowercase","lt_hci","lt_hciexp","lt_qol","lt_hdi"]]
# Preparing a DF for the indices and metrics grouped by region for use in a later analysis
leisuretime_reg = leisuretime_main.groupby("country_region").mean().round(4).sort_values(by="country_region")
leisuretime_reg = leisuretime_reg.reset_index(level=0)
leisuretime_reg = leisuretime_reg.rename(columns={"index":"country_region"})

# CLEANING: Main DF for Market Size (MSi) Metrics
# Cleaning for 2020 Population and 2020 Yearly Change
msi_pop2020 = r_wb_pop2020[["Country (or dependency)",
                           "Population (2020)",
                           "Yearly Change"]]
msi_pop2020 = msi_pop2020.rename(columns={"Country (or dependency)":"country",
                                        "Population (2020)":"msi_pop2020",
                                        "Yearly Change":"msi_change%"})
msi_pop2020["country"] = msi_pop2020["country"].str.strip()
msi_pop2020["msi_change%"] = msi_pop2020["msi_change%"].str.strip()
msi_pop2020["msi_change%"] = msi_pop2020["msi_change%"].str.replace("%","").astype("float")
msi_pop2020["msi_change%"] = (msi_pop2020["msi_change%"]/100).round(4)
msi_pop2020["country_lowercase"] = msi_pop2020["country"].str.lower()
msi_pop2020 = msi_pop2020.drop(columns=["country"])
msi_pop2020 = msi_pop2020.groupby("country_lowercase",as_index=False).mean()
msi_pop2020 = msi_pop2020.drop_duplicates(subset=["country_lowercase"])
# Cleaning file for Age Distribution of Population
msi_agedist = r_wb_agedist[["Country",
                           "Age 15 to 64 Years"]]
msi_agedist = msi_agedist.rename(columns={"Country":"country",
                                        "Age 15 to 64 Years":"msi_midage%"})
msi_agedist["country"] = msi_agedist["country"].str.strip()
msi_agedist["msi_midage%"] = msi_agedist["msi_midage%"].str.strip()
msi_agedist["msi_midage%"] = msi_agedist["msi_midage%"].str.replace("%","").astype("float")
msi_agedist["msi_midage%"] = (msi_agedist["msi_midage%"]/100).round(4)
msi_agedist["country_lowercase"] = msi_agedist["country"].str.lower()
msi_agedist = msi_agedist.drop(columns=["country"])
msi_agedist = msi_agedist.groupby("country_lowercase",as_index=False).mean()
msi_agedist = msi_agedist.drop_duplicates(subset=["country_lowercase"])
# Merging all indicators for Market Size
msi_indicators_df = [msi_pop2020,msi_agedist]
marketsize_main = wb_country
for df in msi_indicators_df:
    marketsize_main = pd.merge(marketsize_main,
                               df,
                               how="left",
                               on="country_lowercase")
# Filling null values using median number per region
msi_indicators_col = ["msi_pop2020","msi_change%","msi_midage%"]
for column in msi_indicators_col:
    marketsize_main[column] = marketsize_main.groupby("country_region")[column].transform(lambda x:x.fillna(x.median()))
    marketsize_main[column] = marketsize_main[column].round(4)
# Calculating 2020 Population specifically for middle-aged (15-64)
marketsize_main["msi_popmid"] = (marketsize_main["msi_pop2020"]*marketsize_main["msi_midage%"]).round(0)
marketsize_main = marketsize_main.drop(columns=["msi_pop2020","msi_midage%"])
# Preparing a trimmed DF for merging later in the analysis
msi_trim = marketsize_main[["country_lowercase","msi_popmid","msi_change%"]]
# Preparing a DF for the indices and metrics grouped by region for use in a later analysis
marketsize_reg = marketsize_main.groupby("country_region").mean().round(4)
marketsize_reg["msi_popmid_000"] = (marketsize_reg["msi_popmid"]/1000).round(0)
marketsize_reg = marketsize_reg.drop(columns=["msi_popmid"])
marketsize_reg = marketsize_reg.reset_index(level=0)
marketsize_reg = marketsize_reg.rename(columns={"index":"country_region"})

# CLEANING: Main DF for Market Stability (MSt) Indices
# Cleaning for Cost of Living and Local Purchasing Power Indices
mst_collpp = r_wb_col[["Country",
                       "Cost of Living Index",
                       "Local Purchasing Power Index"]]
mst_collpp = mst_collpp.rename(columns={"Country":"country",
                                        "Cost of Living Index":"mst_col",
                                        "Local Purchasing Power Index":"mst_lpp"})
mst_collpp["country_lowercase"] = mst_collpp["country"].str.strip().str.lower()
mst_collpp = mst_collpp.drop(columns=["country"])
mst_collpp = mst_collpp.groupby("country_lowercase",as_index=False).mean()
mst_collpp = mst_collpp.drop_duplicates(subset=["country_lowercase"])
marketstab_main = pd.merge(wb_country,
                           mst_collpp,
                           how="left",
                           on="country_lowercase")
# Filling null values using median number per region
mst_indicators_col = ["mst_col","mst_lpp"]
for column in mst_indicators_col:
    marketstab_main[column] = marketstab_main.groupby("country_region")[column].transform(lambda x:x.fillna(x.median()))
    marketstab_main[column] = marketstab_main[column].round(4)
# Preparing a trimmed DF for merging later in the analysis
mst_trim = marketstab_main[["country_lowercase","mst_col","mst_lpp"]]
# Preparing a DF for the indices and metrics grouped by region for use in a later analysis
marketstab_reg = marketstab_main.groupby("country_region").mean().round(4)
marketstab_reg = marketstab_reg.reset_index(level=0)
marketstab_reg = marketstab_reg.rename(columns={"index":"country_region"})
```

## North America - Supply Chain Dataset

After indentifying the results for *Problem 1 - Regional Market Research*, it was necessary to find a dataset in the provided files that would be most indicative of the demand for sports and leisure goods in the United States. The *Supply Chain Dataset (DataCoSupplyChainDataset.csv.zip)* was found to have had the most useful features and observations for this among all datasets related to the US. This database will be used in formulating a strategy for *Problem 2 - Product Strategy*, specifically for North America and the United States.

There were however many unnecessary information and unclean data from the database. As such, the following steps were taken to determine the useful information while retaining data integrity:

- Only relevant columns were included in the main database for analyses. They were all organized and renamed for easier identification.
- Values were cleaned and translated from Spanish to English as necessary. Specifically:
    - Two (2) clearly invalid observations were removed (i.e., State Codes were numerical).
    - "EE. UU." was a convention used to refer to the United States. The replaced by "United States" for clarity.

```python
# NORTH AMERICA

# ACCESSING SUPPLY CHAIN INFORMATION
zf_supply = zf.ZipFile("data/DataCoSupplyChainDataset.csv.zip")
r_supply = pd.read_csv("data/DataCoSupplyChainDataset.csv",low_memory=False,encoding="latin-1")

# CLEANING
# Selecting relevant columns and renaming
na_sales = r_supply[["Product Card Id","Product Name","Category Name",
                     "Customer Id","Customer State","Customer Country","Customer Segment","Department Name",
                     "Market","Order Country","Order Item Product Price", "Order Item Quantity","Sales"]]
na_sales = na_sales.rename(columns={"Product Card Id":"prod_id",
                                    "Product Name":"prod_name",
                                    "Category Name":"cat_name",
                                    "Customer Id":"cus_id",
                                    "Customer State":"cus_state",
                                    "Customer Country":"cus_country",
                                    "Customer Segment":"cus_segment",
                                    "Department Name":"dep_name",
                                    "Market":"market",
                                    "Order Country":"ord_country",
                                    "Order Item Product Price":"price", # same as product price
                                    "Order Item Quantity":"quantity",
                                    "Sales":"sales"}) # price * quantity
# Cleaning values
na_sales = na_sales[na_sales["cus_state"] != "95758"] # deletes 1 invalid observation
na_sales = na_sales[na_sales["cus_state"] != "91732"] # deletes 1 invalid observation
na_sales["cus_country"] = na_sales["cus_country"].replace("EE. UU.","United States") # database used Spanish acronyms, translated to equivalent
na_sales["price"] = na_sales["price"].round(2)

# Binning Prices 
na_sales["price_level"] = pd.cut(na_sales["price"],
                                 bins=5,
                                 labels=("Very Low","Low","Mid","High","Very High"))

# Trimming whitespace and assigning category datatypes
na_sales["prod_name"] = na_sales["prod_name"].str.strip()
na_sales["cat_name"] = na_sales["cat_name"].str.strip().astype("category")
na_sales["cus_state"] = na_sales["cus_state"].str.strip().astype("category")
na_sales["cus_country"] = na_sales["cus_country"].str.strip().astype("category")
na_sales["cus_segment"] = na_sales["cus_segment"].str.strip().astype("category")
na_sales["dep_name"] = na_sales["dep_name"].str.strip().astype("category")
na_sales["market"] = na_sales["market"].str.strip().astype("category")
na_sales["ord_country"] = na_sales["ord_country"].str.strip().astype("category")
na_sales["price_level"] = na_sales["price_level"].astype("category")
```

## Middle East & North Africa - Noon

Similarly, it was necessary to also find a dataset among the provided files to analyze the Middle East & North Africa sports and leisure goods market, specifically for the UAE. The *Noon File (noon.csv)* was found to be the most reasonably related to the UAE market. This database will be helpful in the formulation of a strategic approach for *Problem 2 - Product Strategy*, specifically for the Middle East & North Africa and the starting country for that region - the UAE.

There were however many unnecessary information and unclean data from the database. As such, the following steps were taken to determine the useful information while retaining data integrity:

- Only relevant columns were included in the main database for analyses. They were organized and renamed accordingly, for easier identification.
- Values were cleaned and null values were filled as appropriate.
- A number of categorical features had the number of categories decreased to < 5 for more meaningful analyses. Specifically:
    - Materials were categorized to either Cotton/Cotton Blend, Polyester/Polyester Blend, and Other.
    - Sections were categorized to Men, Women, Unisex and Other.
    - Brands were limited to the top 5 brands with the most products in the dataset and the rest were lumped into Other.

```python
# MIDDLE EAST AND NORTH AFRICA

# ACCESSING THE NOON FILE
r_noon = pd.read_csv("data/noon.csv",low_memory=False)

# CLEANING
# Selecting relevant columns and renaming
me_sales = r_noon[["name","brand","sold_by","section","material","old_price",]]
me_sales = me_sales.rename(columns={"old_price":"price"})
me_sales = me_sales.astype({"brand":"category",
                            "sold_by":"category",
                            "section":"category",
                            "material":"category"})

# Cleaning values, filling nulls, and reducing categories based on relevance
me_sales["mat_clean"] = me_sales["material"].apply(lambda x: "Cotton/Cotton Blend" if "Cotton" in x
                                                     else("Polyester/Polyester Blend" if "Polyester" in x
                                                          else "Other"))
me_sales["mat_clean"] = me_sales["mat_clean"].fillna("Other")
me_sales["sec_clean"] = me_sales["section"].apply(lambda x: "Men" if "Men" in x
                                              else("Women" if "Women" in x
                                                   else("Unisex" if "Unisex" in x
                                                        else "Other")))
me_sales["sec_clean"] = me_sales["sec_clean"].fillna("Other")
me_sales["brand_clean"] = me_sales["brand"].apply(lambda x: "adidas" if "adidas" in x
                                                  else("Nike" if "Nike" in x
                                                       else("Reebok" if "Reebok" in x
                                                            else("UNDER ARMOUR" if "UNDER ARMOUR" in x
                                                                 else("PUMA" if "PUMA" in x
                                                                      else "Other")))))
me_sales["brand_clean"] = me_sales["brand_clean"].fillna("Other")

# Declaring datatypes
me_sales = me_sales.astype({"mat_clean":"category",
                            "sec_clean":"category",
                            "brand_clean":"category"})

# Binning prices
me_sales["price_level"] = pd.cut(me_sales["price"],
                                 bins=5,
                                 labels=("Very Low","Low","Mid","High","Very High"))
```

# Problem Solving

## Problem 1 - Regional Market Research

In determining where the best countries and regions to expand to would be, the following elements were considered as primary criteria of equal importance: **Leisure Time**, **Market Size**, and **Market Stability**. Various economic indices and metrics were then identified and used to serve as indicators for each element. 

**Methodology**
>Each country was ranked and then binned to 5 groups (i.e., 20% per group) to determine how they compare to each other in each index or metric and scored. Groups were scored from 5 to 1 with 5 for those in the highest 20% and 1 for those in the lowest 20%. These points were then weighted depending on how many indices or metrics were used for each of the primary criteria as mentioned above.
>
>All countries were then taken together by their respective regions and the mean of the weighted scores for all countries in that region was calculated per index and then per criteria. All mean scores for each criteria were then summed up to get the total mean score of each region. Among the highest regions, one country was then identified as a starting point for the planned expansion based the total mean score per country.
>
>The analysis was initially done at a regional level to account for the positive impact of ideal countries in proximity of one another. Similar to how the Company was able to establish its brand and operations within European and surrounding countries, it could do similarly to other regions by starting the expansion in the most ideal country within that region. Thereafter, the Company could expand further to surrounding countries.

### Leisure Time

Sports apparel are widely considered to be leisure goods. Specifically, they are leisure goods often purchased in an effort improve health and wellness. In this manner, they can be considered as active leisure goods which are normal goods or goods whose demand rises with health, education and income (Roy & Orazem, 2021).

Healthcare indices were used to account for a potential market's consideration and prioritization of their health. Specifically, the *Health Care Index* is an indicator that estimates the quality of the healthcare system in the country, including the professionals, equipment, and cost of the same among others. Further, the *Healthcare Exponential Index* is another indicator for a country's healthcare system which emphasizes better quality healthcare. (Numbeo, n.d.) The *Human Development Index* was used as well to consider the level of education alongside the health and standard of living. This index was established by the United Nations Development Programme (UNDP) to provide a measure for various key areas of human development. (Sagar & Najam, 1998)

Evidence has suggested that populations with higher levels of income generally have more time to spend on voluntary leisure activities, including sports events. This further impacts the population's well-being and quality of life positively. (Eurostat, 2021) This is reflected in the analysis through the use of the *Quality of Life Index* which is a measure of the social welfare of the people in a country considering various economic indicators as well as a person's personal satisfaction with their lives in the context of their country and culture (Shu et al., 2022).

```python
# Analysis of Leisure Time

# 1. Group ranking the countries based on each indicator on a percentile basis at 5 groups (i.e., using pd.qcut)
# 2. Highest group is given 5 points. Lowest group is given 1 point.
# 3. Leisure Time, Market Size, and Market Stability are assumed to be of equal importance. So points are weighted accordingly.
# Leisure Time has 4 indicators. As such, each indicator is given 25% weight of the weight.

# Group ranking and Weighting of Ranks
lt_list = ["lt_hci","lt_hciexp","lt_qol","lt_hdi"]
leisuretime_r = pd.merge(wb_country,
                         lt_trim,
                         how="left",
                         on="country_lowercase")
for column in lt_list:
    leisuretime_r[f"r_{column}"] = pd.qcut(leisuretime_r[column],
                                           5,
                                           labels=[1,2,3,4,5])
    leisuretime_r[f"r_{column}"] = leisuretime_r[f"r_{column}"].astype("float")
    leisuretime_r[f"wr_{column}"] = leisuretime_r[f"r_{column}"] * 0.25

# Summing the weighted ranks
leisuretime_r["lt_total"] = (leisuretime_r["wr_lt_hci"] +
                             leisuretime_r["wr_lt_hciexp"] +
                             leisuretime_r["wr_lt_qol"] +
                             leisuretime_r["wr_lt_hdi"])

# 4. The countries are grouped per region and the mean score is calculated. 
# The mean is used to avoid the number of countries from impacting the final evaluation of each region.
leisuretime_r_reg = leisuretime_r[["country_region",
                                   "wr_lt_hci",
                                   "wr_lt_hciexp",
                                   "wr_lt_qol",
                                   "wr_lt_hdi",
                                   "lt_total"]]
leisuretime_r_reg = leisuretime_r_reg.groupby("country_region").mean().round(2)
leisuretime_r_reg = leisuretime_r_reg.reset_index(level=0)
leisuretime_r_reg = leisuretime_r_reg.rename(columns={"index":"country_region"})

# Making Visualizations - Categorical chart showing how each region performed for each indicator
leisuretime_r_reg_melt = leisuretime_r_reg.drop(columns=["lt_total"])
leisuretime_r_reg_melt = leisuretime_r_reg_melt.rename(columns={"wr_lt_hci":"Health Care",
                                                                "wr_lt_hciexp":"Health Care Exponential",
                                                                "wr_lt_qol":"Quality of Life",
                                                                "wr_lt_hdi":"Human Development"})
leisuretime_r_reg_melt = pd.melt(leisuretime_r_reg_melt, id_vars=["country_region"])



lt = sns.catplot(
    data=leisuretime_r_reg_melt, 
    y="country_region", x="value", 
    hue="variable", kind="bar",
    height=4,aspect=2)
lt.set_axis_labels("Mean Weighted Score","Region")
lt.despine(left=True)
lt.set(title="Analysis of Leisure Time per Index")
sns.move_legend(
    lt, "lower center",
    bbox_to_anchor=(0.42, -.18),
    ncol=2,
    title="Leisure Time Indices", frameon=True)
```

```python
# Making Visualizations - Categorical chart showing how each region performed in total
leisuretime_r_reg_melt2 = leisuretime_r_reg.drop(columns=["wr_lt_hci","wr_lt_hciexp","wr_lt_qol","wr_lt_hdi"])
leisuretime_r_reg_melt2 = leisuretime_r_reg_melt2.rename(columns={"lt_total":"Total Mean Weighted Rank"})
leisuretime_r_reg_melt2 = pd.melt(leisuretime_r_reg_melt2, id_vars=["country_region"])

lt2 = sns.catplot(
    data=leisuretime_r_reg_melt2,
    y="country_region", x="value",
    hue="variable", kind="bar",
    height=3,aspect=2,legend=False)
lt2.set_axis_labels("Total Mean Weighted Score","Region")
lt2.despine(left=True)
lt2.set(title="Analysis of Overall Leisure Time")
```

**Results for Leisure Time**

>As seen in the figures above, North American countries ranked dominantly above other regions, even above Europe & Central Asia. This was primarily driven by all countries in the region belonging to the top 20% of countries in terms of the *Health Care Exponential Index*, *Quality of Life Index*, and *Human Development Index*. This would mean that the North American market would be the most welcoming to leisure goods.

### Market Size

Of equal importance to how the potential market perceives leisure goods is the size of the market and how this is expected to grow. To estimate the general size of the target market, the 2020 population of each country as provided was multiplied to the estimated percentage of the working population, specifically those aged 15 to 64. This metric for the *Working-aged Population* is intended to represent the size of the possible target market based on the current positioning of the Company's sports apparel products. On the other hand, the *Yearly Change* of the population calculated on the year 2020 was used as the indicator to represent potential market growth.

```python
# Analysis of Market Size

# 1. Group ranking the countries based on each indicator on a percentile basis at 5 groups (i.e., using pd.qcut)
# 2. Highest group is given 5 points. Lowest group is given 1 point.
# 3. Leisure Time, Market Size, and Market Stability are assumed to be of equal importance. So points are weighted accordingly.
# Market Size has 2 indicators. As such, each indicator is given 50% weight of the weight.

# Group ranking and Weighting of Ranks
msi_list = ["msi_popmid","msi_change%"]
marketsize_r = pd.merge(wb_country,
                        msi_trim,
                        how="left",
                        on="country_lowercase")
for column in msi_list:
    marketsize_r[f"r_{column}"] = pd.qcut(marketsize_r[column],
                                           5,
                                           labels=[1,2,3,4,5])
    marketsize_r[f"r_{column}"] = marketsize_r[f"r_{column}"].astype("float")
    marketsize_r[f"wr_{column}"] = marketsize_r[f"r_{column}"] * 0.5

# Summing the weighted ranks
marketsize_r["msi_total"] = (marketsize_r["wr_msi_popmid"] +
                             marketsize_r["wr_msi_change%"])

# 4. The countries are grouped per region and the mean score is calculated. 
# The mean is used to avoid the number of countries from impacting the final evaluation of each region.
marketsize_r_reg = marketsize_r[["country_region",
                                 "wr_msi_popmid",
                                 "wr_msi_change%",
                                 "msi_total"]]
marketsize_r_reg = marketsize_r_reg.groupby("country_region").mean().round(2)
marketsize_r_reg = marketsize_r_reg.reset_index(level=0)
marketsize_r_reg = marketsize_r_reg.rename(columns={"index":"country_region"})

# Making Visualizations - Categorical chart showing how each region performed for each indicator
marketsize_r_reg_melt = marketsize_r_reg.drop(columns=["msi_total"])
marketsize_r_reg_melt = marketsize_r_reg_melt.rename(columns={"wr_msi_popmid":"Working Age Population",
                                                              "wr_msi_change%":"Population Year-on-Year Change",})
marketsize_r_reg_melt = pd.melt(marketsize_r_reg_melt, id_vars=["country_region"])

msi = sns.catplot(
    data=marketsize_r_reg_melt,
    y="country_region", x="value",
    hue="variable", kind="bar",
    height=4,aspect=2)
msi.set_axis_labels("Mean Weighted Score","Region")
msi.despine(left=True)
msi.set(title="Analysis of Market Size per Metric")
sns.move_legend(
    msi, "lower center",
    bbox_to_anchor=(0.42, -.1),
    ncol=2,
    title="Market Size Metrics", frameon=True)
```

```python
# Making Visualizations - Categorical chart showing how each region performed in total
marketsize_r_reg_melt2 = marketsize_r_reg.drop(columns=["wr_msi_popmid","wr_msi_change%"])
marketsize_r_reg_melt2 = marketsize_r_reg_melt2.rename(columns={"msi_total":"Total Mean Weighted Rank"})
marketsize_r_reg_melt2 = pd.melt(marketsize_r_reg_melt2, id_vars=["country_region"])

msi2 = sns.catplot(
    data=marketsize_r_reg_melt2,
    y="country_region", x="value",
    hue="variable", kind="bar",
    height=3,aspect=2,legend=False)
msi2.set_axis_labels("Total Mean Weighted Score","Region")
msi2.despine(left=True)
msi2.set(title="Analysis of Overall Market Size")
```

**Results for Market Size**

> Unlike Leisure Time, North America scored poorly in terms of Market Size. Instead, Sub-Saharan Africa, Middle East & North Africa, and South Asia scored the highest. All three regions scoring high seemed to be largely driven by how their countries scored in the *Year-on-year change* of population by 2020.

### Market Stability

As a normal good, demand increases for leisure goods as the population's general income increases. It however has diminishing marginal increases as income goes up. Further, evidence suggests that as general income of a market increases, the more leisure goods are normalize in the perspective of the population (Houston and Wilson, 2002). In this regard, for a mass-market brand like that of the Company, the countries to expand to must have sufficient market stability wherein the potential population already has the ability to purchase leisure goods.

In this regard, the *Local Purchasing Power Index* was used to account for the ability of a person earning an average wage in the country to purchase basic goods and services in their location (Ali et al., 2022). Additionally, the *Cost of Living Index* was also used to consider the the livability of the country and the income stability of the potential market participants (Porter, 2022). 


```python
# Analysis of Market Stability

# 1. Group ranking the countries based on each indicator on a percentile basis at 5 groups (i.e., using pd.qcut)
# 2. Highest group is given 5 points. Lowest group is given 1 point.
# 3. Leisure Time, Market Size, and Market Stability are assumed to be of equal importance. So points are weighted accordingly.
# Market Stability has 2 indicators. As such, each indicator is given 50% weight of the weight.

# Group ranking and Weighting of Ranks
mst_list = ["mst_col","mst_lpp"]
marketstab_r = pd.merge(wb_country,
                        mst_trim,
                        how="left",
                        on="country_lowercase")
for column in mst_list:
    marketstab_r[f"r_{column}"] = pd.qcut(marketstab_r[column],
                                           5,
                                           labels=[1,2,3,4,5])
    marketstab_r[f"r_{column}"] = marketstab_r[f"r_{column}"].astype("float")
    marketstab_r[f"wr_{column}"] = marketstab_r[f"r_{column}"] * 0.5

# Summing the weighted ranks
marketstab_r["mst_total"] = (marketstab_r["wr_mst_col"] +
                             marketstab_r["wr_mst_lpp"])

# 4. The countries are grouped per region and the mean score is calculated. 
# The mean is used to avoid the number of countries from impacting the final evaluation of each region.
marketstab_r_reg = marketstab_r[["country_region",
                                 "wr_mst_col",
                                 "wr_mst_lpp",
                                 "mst_total"]]
marketstab_r_reg = marketstab_r_reg.groupby("country_region").mean().round(2)
marketstab_r_reg = marketstab_r_reg.reset_index(level=0)
marketstab_r_reg = marketstab_r_reg.rename(columns={"index":"country_region"})

# Making Visualizations - Categorical chart showing how each region performed for each indicator
marketstab_r_reg_melt = marketstab_r_reg.drop(columns=["mst_total"])
marketstab_r_reg_melt = marketstab_r_reg_melt.rename(columns={"wr_mst_col":"Cost of Living",
                                                              "wr_mst_lpp":"Local Purchasing Power",})
marketstab_r_reg_melt = pd.melt(marketstab_r_reg_melt, id_vars=["country_region"])

mst = sns.catplot(
    data=marketstab_r_reg_melt,
    y="country_region", x="value",
    hue="variable", kind="bar",
    height=4,aspect=2)
mst.set_axis_labels("Mean Weighted Score","Region")
mst.despine(left=True)
mst.set(title="Analysis of Market Stability per Index")
sns.move_legend(
    mst, "lower center",
    bbox_to_anchor=(0.42, -.1),
    ncol=2,
    title="Market Stability Indices", frameon=True)
```

```python
# Making Visualizations - Categorical chart showing how each region performed in total
marketstab_r_reg_melt2 = marketstab_r_reg.drop(columns=["wr_mst_col","wr_mst_lpp"])
marketstab_r_reg_melt2 = marketstab_r_reg_melt2.rename(columns={"mst_total":"Total Mean Weighted Rank"})
marketstab_r_reg_melt2 = pd.melt(marketstab_r_reg_melt2, id_vars=["country_region"])

mst2 = sns.catplot(
    data=marketstab_r_reg_melt2,
    y="country_region", x="value",
    hue="variable", kind="bar",
    height=3,aspect=2,legend=False)
mst2.set_axis_labels("Total Mean Weighted Score","Region")
mst2.despine(left=True)
mst2.set(title="Analysis of Overall Market Stability")
```

## Problem 2 - Product Strategy for UAE

The database used to approximate the demand for sporting and leisure goods in the UAE provides information about the brands, sections, and types of materials used for the products. To streamline the analyses specifically for the UAE, the below goals were built around the available information the database used could provide insight to. The following analyses will be focused on these goals and together they should be able to provide a well-founded and data-driven product strategy for the Company.
1. Identify the best local UAE distributors to establish relationships with;
2. Identify the brands to potentially compete with and the corresponding pricing levels to be mindful of; and
3. Identify the sections with a potential market gap.

```python
# Analyzing Middle East & North America Markets through the UAE
me_sales_uae = me_sales[["name",
                         "brand_clean",
                         "sold_by",
                         "sec_clean",
                         "mat_clean",
                         "price",
                         "price_level"]]
me_sales_uae["count"] = me_sales_uae["price"].apply(lambda x: 1 if x != 0 else 0)

# Making Visualizations - Sunburst chart showing division between distributors and brands
me_sales_dis = me_sales_uae[["brand_clean",
                             "sold_by",
                             "count"]]
me_sales_dis = me_sales_dis.groupby(["sold_by","brand_clean"],as_index=False)["count"].sum()
me_sales_dis = me_sales_dis[me_sales_dis["count"] >= 25]

fig_uae_dis = px.sunburst(data_frame=me_sales_dis,
                         path=["sold_by","brand_clean"],
                         values="count")

fig_uae_dis.update_layout(autosize=False,
                  width=550,
                  height=550,
                  margin=dict(l=30,
                              r=30,
                              b=80,
                              t=80,
                              pad=4),
                  title_text = 'Distributor and Brands with >25 sales in the UAE')
fig_uae_dis.show()
```

**Distributor and Corresponding Brands**
> It is clear in the above figure that *noon* is the more dominant distributor in the UAE. Nonetheless, Reebook, Nike and adidas have established relationships with both *noon* and *Namshi* for their goods to be sold therein. In this regard, it would be important for the Company to establish a good relationship with *noon* first but also ensure that they can establish presence in *Namshi* as well. As the Company enters the market, they need to be visible alongside their competitors to be seen at the same level.

```python
# Making Visualizations - Horizontal violin chart showing brands, distributors and pricing levels
me_sales_pr = me_sales_uae[["sold_by",
                            "brand_clean",
                            "price"]]
me_sales_pr = me_sales_uae[me_sales_uae["sold_by"] != "SIVVI"]
me_sales_pr["sold_by"] = me_sales_pr["sold_by"].cat.remove_unused_categories()
me_sales_pr = me_sales_pr[(np.abs(stats.zscore(me_sales_pr["price"])) < 3)]
me_sales_pr_c = sns.violinplot(data=me_sales_pr,
               x="price", y="brand_clean",hue="sold_by",
               split=True)
sns.move_legend(
    me_sales_pr_c, "lower center",
    bbox_to_anchor=(0.4, -.3),
    ncol=3,
    title="Distributor", frameon=True)
```

**Competing Brands and Pricing Levels**
> The above figure shows that all possible competing sports apparel brands are pricing around the same levels, particularly Nike, adidas, and Reebok. The above can also serve as a rough guide as to how the Company should potentially price their products.

```python
# Making Visualizations - Bar graphs sections per material
me_sales_mat = me_sales_uae[["brand_clean",
                             "sold_by",
                             "sec_clean",
                             "mat_clean",
                             "price",
                             "count"]]
me_sales_mat = me_sales_mat.groupby(["sec_clean","mat_clean"],as_index=False)["count"].sum()
me_sales_mat = me_sales_mat[me_sales_mat["count"] >= 50]
me_sales_mat["sec_clean"] = me_sales_mat["sec_clean"].cat.remove_unused_categories()

me_sales_mat_c= sns.catplot(
    data=me_sales_mat, x="count", y="mat_clean", hue="sec_clean",
    kind="bar", height=2.5, aspect=3,
)
sns.move_legend(
    me_sales_mat_c, "lower center",
    bbox_to_anchor=(0.45, -.2),
    ncol=3,
    title="Section", frameon=True)
```

**Product Sections and Materials**
> The above figure shows a clear gap for sports apparel for women made from Cotton/Cotton Blend. The Company could take advantage of this market gap and together with all the other information above about distributors and pricing, can penetrate the UAE market more efficiently and effectively.
