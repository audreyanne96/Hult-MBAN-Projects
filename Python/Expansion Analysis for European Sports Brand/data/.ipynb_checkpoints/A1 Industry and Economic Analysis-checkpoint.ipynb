```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile as zf
```

```python
# World Bank Info
# Using 2020 info as much as possible

zf_wb = zf.ZipFile("World Bank.zip")

# Countries and Regions
r_wb_country = pd.read_csv(zf_wb.open("world-development-indicators/Country.csv"),low_memory=False) 

# Indexes = Cost of Living, Rent, CoL+Rent, Groceries, Restaurant, Local Purchasing Power, McMeal
r_wb_col = pd.read_csv(zf_wb.open("2020-cost-of-living/cost of living 2020.csv"),low_memory=False) 
    #135 countries, first Switzerland at 122.40, START of year per google

# Index = Healthcare
r_wb_healthcare = pd.read_csv(zf_wb.open("countries-dataset-2020/Health care index by countries 2020.csv"),low_memory=False) 
    #93 countries, first Taiwan at 86.71, START of year per google search
    
# Index = Quality of Life, Purchasing Power, Safety Index, Health Care Index, CoL, Property Price to Income, Traffic Commute Time, Pollution, Climate
r_wb_qol = pd.read_csv(zf_wb.open("countries-dataset-2020/Quality of life index by countries 2020.csv"),low_memory=False)
    #80 Countries, QoL highest Denmark at 192.67, START of year, matches start of year CoL
    
# Population = change in population overtime
r_wb_popovertime = pd.read_csv(zf_wb.open("country_population.csv"),low_memory=False)
    # until 2016, with countries and regions
    
# Population = change in population overtime
r_wb_pop2020 = pd.read_csv(zf_wb.open("population-by-country-2020/population_by_country_2020.csv"),low_memory=False)
    # 2020, pop, yearly change, net change, density, land area, migrants, fert. rate, med age, urban pop

# Population Density
r_wb_popdens = pd.read_csv(zf_wb.open("countries-dataset-2020/Pupulation density by countries.csv"),low_memory=False)
    # area in km, area in mi, population, density in km, density in mi

# Developed/Developing/Under-developed
r_wb_deved = pd.read_csv(zf_wb.open("list_of_developed_countries.csv"),low_memory=False)
r_wb_deving = pd.read_csv(zf_wb.open("list_of_developed_or_developing_countries.csv"),low_memory=False)
r_wb_udev = pd.read_csv(zf_wb.open("list_of_less_or_under_developed_countries.csv"),low_memory=False)

# HH Exp
r_wb_hhexp = pd.read_csv(zf_wb.open("householdexpenditure/householdexpenditure.csv"),low_memory=False)
```

```python
# Cleaning Countries and Regions
wb_country = r_wb_country[["Alpha2Code",
                           "ShortName",
                           "LongName",
                           "Region"]]
wb_country = wb_country.rename(columns={"Alpha2Code":"country_code",
                                        "ShortName":"country",
                                        "LongName":"country_full",
                                        "Region":"country_region"})
wb_country = wb_country.dropna(subset=["country_region"]) # Dropping rows related to aggregations which have no associated regions
wb_country["country_code"] = wb_country["country_code"].str.strip()
wb_country["country"] = wb_country["country"].str.strip()
wb_country["country_full"] = wb_country["country_full"].str.strip()
wb_country["country_region"] = wb_country["country_region"].str.strip()
wb_country["country_region"] = wb_country["country_region"].astype("category")
wb_country["country_lowercase"] = wb_country["country"].str.lower()
wb_country.info()
```

```python
# Cleaning Cost of Living
wb_col = r_wb_col[["Country",
                   "Cost of Living Index"]]
wb_col = wb_col.rename(columns={"Country":"country",
                                "Cost of Living Index":"i_costofliving"})
wb_col["country_lowercase"] = wb_col["country"].str.strip().str.lower()
wb_col = wb_col.drop(columns=["country"])
wb_col = wb_col.drop_duplicates(subset=["country_lowercase"])

# Cleaning Local Purchasing Power
wb_lpp = r_wb_col[["Country",
                   "Local Purchasing Power Index"]]
wb_lpp = wb_lpp.rename(columns={"Country":"country",
                                "Local Purchasing Power Index":"i_localpurchpower"})
wb_lpp["country_lowercase"] = wb_lpp["country"].str.strip().str.lower()
wb_lpp = wb_lpp.drop(columns=["country"])
wb_lpp = wb_lpp.drop_duplicates(subset=["country_lowercase"])

# Cleaning Health Care Index
wb_hci = r_wb_healthcare[["Country",
                          "Health Care Index"]]
wb_hci = wb_hci.rename(columns={"Country":"country",
                                "Health Care Index":"i_healthcare"})
wb_hci["country_lowercase"] = wb_hci["country"].str.strip().str.lower()
wb_hci = wb_hci.drop(columns=["country"])
wb_hci = wb_hci.drop_duplicates(subset=["country_lowercase"])

# Cleaning Quality of Life
wb_qol = r_wb_qol[["Country",
                   "Quality of Life Index"]]
wb_qol = wb_qol.rename(columns={"Country":"country",
                                "Quality of Life Index":"i_qualityoflife"})
wb_qol["country_lowercase"] = wb_qol["country"].str.strip().str.lower()
wb_qol = wb_qol.drop(columns=["country"])
wb_qol = wb_qol.drop_duplicates(subset=["country_lowercase"])

# Cleaning Income Groups
wb_incgroup = r_wb_country[["ShortName",
                            "IncomeGroup"]]
wb_incgroup = wb_incgroup.rename(columns={"ShortName":"country",
                                          "IncomeGroup":"incomegroup"})
wb_incgroup["incomegroup"] = wb_incgroup["incomegroup"].str.strip().astype("category")
wb_incgroup["country_lowercase"] = wb_incgroup["country"].str.strip().str.lower()
wb_incgroup = wb_incgroup.drop(columns=["country"])
wb_incgroup = wb_incgroup.drop_duplicates(subset=["country_lowercase"])

# Dev Tables
wb_dev = pd.concat([r_wb_deved,r_wb_deving,r_wb_udev])
wb_dev = wb_dev.rename(columns={"Country Name":"country",
                                "GNI per capita":"gni",
                                "Human Development Index":"i_humandev",
                                "Human Asset Index":"i_humanasset",
                                "Population 2020":"population2020",
                                "Development Status":"devstat"})
wb_dev["country_lowercase"] = wb_dev["country"].str.strip().str.lower()
wb_dev = wb_dev.drop(columns=["country"])

wb_dev1 = wb_dev[["country_lowercase","devstat","gni"]]
wb_dev2 = wb_dev.groupby("country_lowercase",as_index=False).mean()

# GNI
wb_gni = wb_dev1[["country_lowercase",
                  "gni"]]
wb_gni["gni"] = wb_gni["gni"].fillna("0").map(lambda x:x.strip("USD").strip().replace(",","").replace("-","0")).astype("float")
wb_gni = wb_gni.groupby("country_lowercase",as_index=False).mean()
wb_gni = wb_gni.drop_duplicates(subset=["country_lowercase"])

# HDI
wb_hdi = wb_dev2[["country_lowercase",
                  "i_humandev"]]
wb_hdi = wb_hdi.dropna(subset=["i_humandev"])
wb_hdi = wb_hdi.drop_duplicates(subset=["country_lowercase"])

# Development Status
wb_devstat = wb_dev1[["country_lowercase",
                      "devstat"]]
wb_devstat["devstat"] = wb_devstat["devstat"].str.strip().astype("category")
wb_devstat = wb_devstat.dropna(subset=["devstat"])
wb_devstat = wb_devstat.drop_duplicates(subset=["country_lowercase"])

# Household Exp
wb_hhexp = r_wb_hhexp[["Country",
                       "HouseholdExpenditurePerCapita"]]
wb_hhexp = wb_hhexp.rename(columns={"Country":"country",
                                    "HouseholdExpenditurePerCapita":"hhexpenditurepercap"})
wb_hhexp["country_lowercase"] = wb_hhexp["country"].str.strip().str.lower()
wb_hhexp = wb_hhexp.drop(columns=["country"])
wb_hhexp = wb_hhexp.dropna(subset=["hhexpenditurepercap"])
wb_hhexp = wb_hhexp.drop_duplicates(subset=["country_lowercase"])

indicators = [wb_col,wb_lpp,wb_hci,wb_qol,wb_incgroup,wb_gni,wb_hdi,wb_devstat,wb_hhexp]
wb_main = wb_country

for df in indicators:
    wb_main = pd.merge(wb_main,
                       df,
                       how="left",
                       on="country_lowercase")

wb_main.head(10)
```
