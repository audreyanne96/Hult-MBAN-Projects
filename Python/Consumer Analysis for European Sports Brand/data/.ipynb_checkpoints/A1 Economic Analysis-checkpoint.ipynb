```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import zipfile as zf
```

# **Criteria for Region Expansion**

### Leisure Time - countries prioritizing health and sports - Higher the better
    1. HC - Healthcare Index - an estimation of the overall quality of the health care system, health care professionals, etc.
    2. HC - Healthcare Exp IndexÂ - show healthcare index such that it raises more if the health care system is of better quality (measures growth).
    3. QOL - Human Development Index - index of life expectancy, education, & per capita income, used to rank countries into 4 tiers of human development.
    4. QOL - Quality of Life Index - measures quality of life in terms of both satisfaction and importance regarding various aspects of life.
### Market Size - Potential market - Higher the better
    1. Population of middle-aged
    2. Change in Population yoy
### Market Stability - Stability of the population in being able to purchase - Higher the better
    1. Local Purchasing Power 
    2. Cost of Living Index - compare the expenses an average person can expect to incur to acquire food, shelter, transportation, energy, clothing, education, healthcare, childcare, and entertainment.

```python
# World Bank Info
# Using World Bank info because: It has the most information for all countries. Other industry and economic information provided are biased towards US. It is also likely the most reliable.
# Data chosen to be used are 2020 data as identified. If not identified by date, assumed to be 2020. 
# When faced with multiple databases of the same data providing different information, the dataset used is the one found to be more consistent with the rest of the data.

# Opening the zip file for World Bank data
zf_wb = zf.ZipFile("World Bank.zip")

# Countries and Regions
# Chosen to be used since appeared to be the cleanest and most organized.
# Also had the most reasonable division of regions which is not necessarily based on geographic continent.
r_wb_country = pd.read_csv(zf_wb.open("world-development-indicators/Country.csv"),low_memory=False)

# For Healthcare Indices to be used in determining Leisure Time
r_wb_healthcare = pd.read_csv(zf_wb.open("countries-dataset-2020/Health care index by countries 2020.csv"),low_memory=False)

# For Quality of Life Index to be used in determining Leisure Time
r_wb_qol = pd.read_csv(zf_wb.open("countries-dataset-2020/Quality of life index by countries 2020.csv"),low_memory=False)

# For Human Development Index to be used in determining Leisure Time
# Used over other datasets since per checking this set had the most countries with HDIs.
# Assumed 2020 since population information included was from the year 2020
r_wb_deved = pd.read_csv(zf_wb.open("list_of_developed_countries.csv"),low_memory=False)
r_wb_deving = pd.read_csv(zf_wb.open("list_of_developed_or_developing_countries.csv"),low_memory=False)
r_wb_udev = pd.read_csv(zf_wb.open("list_of_less_or_under_developed_countries.csv"),low_memory=False)

# For 2020 population and year-on-year change to be used in determining Market Size
# Used over other datasets since had to most collective information for all countries.
r_wb_pop2020 = pd.read_csv(zf_wb.open("population-by-country-2020/population_by_country_2020.csv"),low_memory=False)

# For % of middle-aged (15-64) population to be used in determining Market Size
r_wb_agedist = pd.read_csv(zf_wb.open("countries-dataset-2020/Coutries age structure.csv"),low_memory=False)

# For Local Purchasing Power Index and Cost of Living Index to be used in determining Market Stability
r_wb_col = pd.read_csv(zf_wb.open("2020-cost-of-living/cost of living 2020.csv"),low_memory=False)

# Cleaning Countries and Regions
wb_country = r_wb_country[["Alpha2Code",
                           "ShortName",
                           "LongName",
                           "Region"]]
wb_country = wb_country.rename(columns={"Alpha2Code":"country_code",
                                        "ShortName":"country",
                                        "LongName":"country_full",
                                        "Region":"country_region"})
wb_country = wb_country.dropna(subset=["country_region"]) # Dropping rows related to aggregations and so have no associated regions
wb_country["country_code"] = wb_country["country_code"].str.strip()
wb_country["country"] = wb_country["country"].str.strip()
wb_country["country_full"] = wb_country["country_full"].str.strip()
wb_country["country_region"] = wb_country["country_region"].str.strip().astype("category")
wb_country["country_lowercase"] = wb_country["country"].str.lower()
```

## Leisure Time - countries prioritizing health and sports - Higher the better

1. HCI - Healthcare Index - an estimation of the overall quality of the health care system, health care professionals, etc.
2. HCI - Healthcare Exp IndexÂ - show healthcare index such that it raises more if the health care system is of better quality (measures growth).
3. QOL - Quality of Life Index - measures quality of life in terms of both satisfaction and importance regarding various aspects of life.
4. QOL - Human Development Index - index of life expectancy, education, & per capita income, used to rank countries into 4 tiers of human development.

```python
# Cleaning Health Care Indices (HCI and HCI Exp)
lt_hci = r_wb_healthcare[["Country",
                          "Health Care Index",
                          "Health Care Exp. Index"]]
lt_hci = lt_hci.rename(columns={"Country":"country",
                                "Health Care Index":"lt_hci",
                                "Health Care Exp. Index":"lt_hciexp"})
lt_hci["country_lowercase"] = lt_hci["country"].str.strip().str.lower()
lt_hci = lt_hci.drop(columns=["country"])
lt_hci = lt_hci.groupby("country_lowercase",as_index=False).mean() # Averaging duplicate entries
lt_hci = lt_hci.drop_duplicates(subset=["country_lowercase"])

# Cleaning Quality of Life Index
lt_qol = r_wb_qol[["Country",
                   "Quality of Life Index"]]
lt_qol = lt_qol.rename(columns={"Country":"country",
                                "Quality of Life Index":"lt_qol"})
lt_qol["country_lowercase"] = lt_qol["country"].str.strip().str.lower()
lt_qol = lt_qol.drop(columns=["country"])
lt_qol = lt_qol.groupby("country_lowercase",as_index=False).mean() # Averaging duplicate entries
lt_qol = lt_qol.drop_duplicates(subset=["country_lowercase"])

# Cleaning Development Tables with HDI
lt_hdi = pd.concat([r_wb_deved,r_wb_deving,r_wb_udev])
lt_hdi = lt_hdi.rename(columns={"Country Name":"country",
                                "Human Development Index":"lt_hdi"})
lt_hdi["country_lowercase"] = lt_hdi["country"].str.strip().str.lower()
lt_hdi = lt_hdi.drop(columns=["country",
                              "GNI per capita",
                              "Human Asset Index",
                              "Population 2020",
                              "Development Status"])
lt_hdi = lt_hdi.groupby("country_lowercase",as_index=False).mean() # Averaging duplicate entries
lt_hdi = lt_hdi.drop_duplicates(subset=["country_lowercase"])

# Combining all indicators for Leisure Time
lt_indicators_df = [lt_hci,lt_qol,lt_hdi]
leisuretime_main = wb_country

for df in lt_indicators_df:
    leisuretime_main = pd.merge(leisuretime_main,
                           df,
                           how="left",
                           on="country_lowercase")

# Filling null values using median number per region
lt_indicators_col = ["lt_hci","lt_hciexp","lt_qol","lt_hdi"]

for column in lt_indicators_col:
    leisuretime_main[column] = leisuretime_main.groupby("country_region")[column].transform(lambda x:x.fillna(x.median()))
    leisuretime_main[column] = leisuretime_main[column].round(4)

# Making trimmed table in preparation for merging later
lt_trim = leisuretime_main[["country_lowercase","lt_hci","lt_hciexp","lt_qol","lt_hdi"]]

# Ranking per region for Lesiure Time
leisuretime_reg = leisuretime_main.groupby("country_region").mean().round(4).sort_values(by="country_region")
leisuretime_reg = leisuretime_reg.reset_index(level=0)
leisuretime_reg = leisuretime_reg.rename(columns={"index":"country_region"})

sns.set_palette("pastel")
sns.barplot(x="lt_hci",y="country_region",data=leisuretime_reg) #lt_hci
sns.barplot(x="lt_hciexp",y="country_region",data=leisuretime_reg) #lt_hciexp
sns.barplot(x="lt_qol",y="country_region",data=leisuretime_reg) #lt_qol
sns.barplot(x="lt_hdi",y="country_region",data=leisuretime_reg) #lt_hdi
```

## 2. Market Size - Potential market - Higher the better

1. Population of middle-aged
2. Change in Population yoy

```python
# Cleaning 2020 Population Data
msi_pop2020 = r_wb_pop2020[["Country (or dependency)",
                           "Population (2020)",
                           "Yearly Change"]]
msi_pop2020 = msi_pop2020.rename(columns={"Country (or dependency)":"country",
                                        "Population (2020)":"msi_pop2020",
                                        "Yearly Change":"msi_change%"})
msi_pop2020["country"] = msi_pop2020["country"].str.strip()
msi_pop2020["msi_change%"] = msi_pop2020["msi_change%"].str.strip()
msi_pop2020["msi_change%"] = msi_pop2020["msi_change%"].str.replace("%","").astype("float")
msi_pop2020["msi_change%"] = (msi_pop2020["msi_change%"]/100).round(4)
msi_pop2020["country_lowercase"] = msi_pop2020["country"].str.lower()
msi_pop2020 = msi_pop2020.drop(columns=["country"])
msi_pop2020 = msi_pop2020.groupby("country_lowercase",as_index=False).mean() # Averaging duplicate entries
msi_pop2020 = msi_pop2020.drop_duplicates(subset=["country_lowercase"])

# Cleaning Population Age Distribution Data
msi_agedist = r_wb_agedist[["Country",
                           "Age 15 to 64 Years"]]
msi_agedist = msi_agedist.rename(columns={"Country":"country",
                                        "Age 15 to 64 Years":"msi_midage%"})
msi_agedist["country"] = msi_agedist["country"].str.strip()
msi_agedist["msi_midage%"] = msi_agedist["msi_midage%"].str.strip()
msi_agedist["msi_midage%"] = msi_agedist["msi_midage%"].str.replace("%","").astype("float")
msi_agedist["msi_midage%"] = (msi_agedist["msi_midage%"]/100).round(4)
msi_agedist["country_lowercase"] = msi_agedist["country"].str.lower()
msi_agedist = msi_agedist.drop(columns=["country"])
msi_agedist = msi_agedist.groupby("country_lowercase",as_index=False).mean() # Averaging duplicate entries
msi_agedist = msi_agedist.drop_duplicates(subset=["country_lowercase"])

# Combining all indicators for Leisure time in one table
msi_indicators_df = [msi_pop2020,msi_agedist]
marketsize_main = wb_country

for df in msi_indicators_df:
    marketsize_main = pd.merge(marketsize_main,
                               df,
                               how="left",
                               on="country_lowercase")

# Filling null values using median number per region
msi_indicators_col = ["msi_pop2020","msi_change%","msi_midage%"]

for column in msi_indicators_col:
    marketsize_main[column] = marketsize_main.groupby("country_region")[column].transform(lambda x:x.fillna(x.median()))
    marketsize_main[column] = marketsize_main[column].round(4)

# Calculate Pop2020 for MidAge 15-64
marketsize_main["msi_popmid"] = (marketsize_main["msi_pop2020"]*marketsize_main["msi_midage%"]).round(0)
marketsize_main = marketsize_main.drop(columns=["msi_pop2020","msi_midage%"])

# Making trimmed table in preparation for merging later
msi_trim = marketsize_main[["country_lowercase","msi_popmid","msi_change%"]]

# per region for Market Size
marketsize_reg = marketsize_main.groupby("country_region").mean().round(4)
marketsize_reg["msi_popmid_000"] = (marketsize_reg["msi_popmid"]/1000).round(0)
marketsize_reg = marketsize_reg.drop(columns=["msi_popmid"])
marketsize_reg = marketsize_reg.reset_index(level=0)
marketsize_reg = marketsize_reg.rename(columns={"index":"country_region"})

sns.barplot(x="msi_change%",y="country_region",data=marketsize_reg) #msi_change%
sns.barplot(x="msi_popmid_000",y="country_region",data=marketsize_reg) #msi_popmid_000
```

## 3. Market Stability - Stability of the population in being able to purchase - Higher the better

1. Local Purchasing Power
2. Cost of Living Index - compare the expenses an average person can expect to incur to acquire food, shelter, transportation, energy, clothing, education, healthcare, childcare, and entertainment.

```python
# Cleaning for Cost of Living and Local Purchasing Power Indices
mst_collpp = r_wb_col[["Country",
                       "Cost of Living Index",
                       "Local Purchasing Power Index"]]
mst_collpp = mst_collpp.rename(columns={"Country":"country",
                                        "Cost of Living Index":"mst_col",
                                        "Local Purchasing Power Index":"mst_lpp"})
mst_collpp["country_lowercase"] = mst_collpp["country"].str.strip().str.lower()
mst_collpp = mst_collpp.drop(columns=["country"])
mst_collpp = mst_collpp.groupby("country_lowercase",as_index=False).mean() # Averaging duplicate entries
mst_collpp = mst_collpp.drop_duplicates(subset=["country_lowercase"])

# Combining all indicators for Leisure time in one table
marketstab_main = pd.merge(wb_country,
                           mst_collpp,
                           how="left",
                           on="country_lowercase")

mst_indicators_col = ["mst_col","mst_lpp"]
for column in mst_indicators_col:
    marketstab_main[column] = marketstab_main.groupby("country_region")[column].transform(lambda x:x.fillna(x.median()))
    marketstab_main[column] = marketstab_main[column].round(4)

# Making trimmed table in preparation for merging later
mst_trim = marketstab_main[["country_lowercase","mst_col","mst_lpp"]]

# Ranking per region for Market Size
marketstab_reg = marketstab_main.groupby("country_region").mean().round(4)
marketstab_reg = marketstab_reg.reset_index(level=0)
marketstab_reg = marketstab_reg.rename(columns={"index":"country_region"})

sns.barplot(x="mst_col",y="country_region",data=marketstab_reg) #mst_col
sns.barplot(x="mst_lpp",y="country_region",data=marketstab_reg) #mst_lpp
```

```python
# Merging all 3 tables of indicators using the prepared trimmed tables for ranking and analyses.
for_merging_tables = [lt_trim,msi_trim,mst_trim]
merged_main = wb_country

for df in for_merging_tables:
    merged_main = pd.merge(merged_main,
                           df,
                           how="left",
                           on="country_lowercase")

# Phase 1:
# 1. Ranking the countries based on each indicator on a percentile basis at 5 groups (i.e., using pd.qcut)
# 2. Highest group is given 5 points. Lowest group is given 1 point.
# 3. Leisure Time, Market Size, and Market Stability are assumed to be of equal importance. So points are weighted accordingly:
# 3a. Leisure Time = 4 indicators = 25% weight per indicator
# 3b. Market Size = 2 indicators = 5% weight per indicator
# 3c. Market Stability = 2 indicators = 5% weight per indicator
for_cutting_columns = ["lt_hci","lt_hciexp","lt_qol","lt_hdi","msi_popmid","msi_change%","mst_col","mst_lpp"]

for column in for_cutting_columns:
    merged_main[f"r_{column}"] = pd.qcut(merged_main[column],
                                             5,
                                             labels=[1,2,3,4,5])
    merged_main[f"r_{column}"] = merged_main[f"r_{column}"].astype("float")
    if column == "lt_hci" or column == "lt_hciexp" or column == "lt_qol" or column == "lt_hdi":
        merged_main[f"wr_{column}"] = merged_main[f"r_{column}"] * 0.25
    elif column == "msi_popmid" or column == "msi_change%" or column == "mst_col" or column == "mst_lpp":
        merged_main[f"wr_{column}"] = merged_main[f"r_{column}"] * 0.5
    else:
        print("error")
        break

# Phase 2:
# 5. The total points are calculated per indicator, per country.
# 6. The total points are calculated per country.
merged_main["lt_total"] = merged_main["wr_lt_hci"] + merged_main["wr_lt_hciexp"] + merged_main["wr_lt_qol"] + merged_main["wr_lt_hdi"]
merged_main["msi_total"] = merged_main["wr_msi_popmid"] + merged_main["wr_msi_change%"]
merged_main["mst_total"] = merged_main["wr_mst_col"] + merged_main["wr_mst_lpp"]
merged_main["total"] = merged_main["lt_total"] + merged_main["msi_total"] + merged_main["mst_total"]


# Phase 3:
# 7. The countries are grouped per region and the mean score is calculated. The mean is used to avoid the number of countries from impacting the final evaluation of each region.
merged_main_reg = merged_main[["country_region",
                               "lt_total",
                               "msi_total",
                               "mst_total",
                               "total"]]
merged_main_reg = merged_main_reg.groupby("country_region").mean().round(2)
merged_main_reg = merged_main_reg.reset_index(level=0)
merged_main_reg = merged_main_reg.rename(columns={"index":"country_region"})


sns.barplot(x="lt_total",y="country_region",data=merged_main_reg) #mst_lpp
sns.barplot(x="msi_total",y="country_region",data=merged_main_reg) #mst_lpp
sns.barplot(x="mst_total",y="country_region",data=merged_main_reg) #mst_lpp
sns.barplot(x="total",y="country_region",data=merged_main_reg) #total
```

```python
# NORTH AMERICA
zf_supply = zf.ZipFile("DataCoSupplyChainDataset.csv.zip")
r_supply = pd.read_csv("DataCoSupplyChainDataset.csv",low_memory=False,encoding="latin-1")

# Cleaning assumed NA Sales

# Selecting relevant columns and renaming
na_sales = r_supply[["Product Card Id","Product Name","Category Name",
                     "Customer Id","Customer State","Customer Country","Customer Segment","Department Name",
                     "Market","Order Country","Order Item Product Price", "Order Item Quantity","Sales"]]
na_sales = na_sales.rename(columns={"Product Card Id":"prod_id",
                                    "Product Name":"prod_name",
                                    "Category Name":"cat_name",
                                    "Customer Id":"cus_id",
                                    "Customer State":"cus_state",
                                    "Customer Country":"cus_country",
                                    "Customer Segment":"cus_segment",
                                    "Department Name":"dep_name",
                                    "Market":"market",
                                    "Order Country":"ord_country",
                                    "Order Item Product Price":"price", # same as product price
                                    "Order Item Quantity":"quantity",
                                    "Sales":"sales"}) # price * quantity

# Cleaning values
na_sales = na_sales[na_sales["cus_state"] != "95758"]
na_sales = na_sales[na_sales["cus_state"] != "91732"]
na_sales["cus_country"] = na_sales["cus_country"].replace("EE. UU.","United States")
na_sales["price"] = na_sales["price"].round(2)

# Binning Prices
na_sales["price_level"] = pd.cut(na_sales["price"],
                                 bins=5,
                                 labels=("very high","high","mid","low","very low"))

# Trimming whitespace and assigning category datatypes
na_sales["prod_name"] = na_sales["prod_name"].str.strip()
na_sales["cat_name"] = na_sales["cat_name"].str.strip().astype("category")
na_sales["cus_state"] = na_sales["cus_state"].str.strip().astype("category")
na_sales["cus_country"] = na_sales["cus_country"].str.strip().astype("category")
na_sales["cus_segment"] = na_sales["cus_segment"].str.strip().astype("category")
na_sales["dep_name"] = na_sales["dep_name"].str.strip().astype("category")
na_sales["market"] = na_sales["market"].str.strip().astype("category")
na_sales["ord_country"] = na_sales["ord_country"].str.strip().astype("category")
na_sales["price_level"] = na_sales["price_level"].astype("category")
```
